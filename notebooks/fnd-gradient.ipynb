# The gradient (slope in many directions)

_Foundations: Math you need first_

**A derivative for functions with many inputs. It points straight uphill.**

Most ML (Machine Learning) functions have many inputs (millions of weights), not one.

     The gradient collects the slope in each input direction into one vector.

---

This notebook builds up the gradient one step at a time. Run each cell, read the note above it, and watch how stepping opposite the gradient walks a point steadily downhill toward the minimum. _Save a copy to your Drive (File → Save a copy in Drive) to edit and keep your work._

```python
# Setup — numpy / pandas / scikit-learn / matplotlib ship with Colab.
import numpy as np
import matplotlib.pyplot as plt
```

## Reference implementation — NumPy

### Step 1 — Define the bowl and its gradient

Our function `f` is a round bowl, `x² + y²`, whose lowest point sits at the origin `(0, 0)`. The gradient of this bowl is `[2x, 2y]`, which is just `2 * p`. The gradient points **straight uphill**, so at the point `[3, 4]` it tells us the steepest-ascent direction and how steep it is.

```python
def f(p):                       # round bowl: x^2 + y^2
    return p[0] ** 2 + p[1] ** 2

def grad(p):                    # gradient = [2x, 2y]
    return 2 * p

p = np.array([3.0, 4.0])
g = grad(p)
print("point   :", p, " f =", f(p))
print("grad    :", g, " (points straight uphill)")
```

### Step 2 — Step opposite the gradient to go downhill

Since the gradient points uphill, subtracting it walks us downhill. Each iteration takes a small step of size `lr` (the learning rate) in the `-gradient` direction. Watch `f` shrink each step as the point slides toward the minimum at `(0, 0)`.

```python
# gradient descent: step opposite the gradient to go downhill
lr = 0.1
for step in range(5):
    p = p - lr * grad(p)        # -grad heads toward the minimum (0,0)
    print("step %d  point=%s  f=%.4f" % (step + 1, np.round(p, 4), f(p)))

print("heading toward min at [0, 0]")
```

## Visualize the data & results

_Does stepping opposite the gradient walk downhill toward the minimum?_

### Step 1 — Record the descent path

To plot the journey we re-run the same gradient descent from `(3, 4)`, this time saving every point we visit and the function value there. `path` collects the (x, y) positions and `losses` collects the matching `f` values so we can chart both.

```python
# gradient descent on f = x^2 + y^2 from (3, 4)
p = np.array([3.0, 4.0])
lr = 0.1
path = [p.copy()]
losses = [float(p @ p)]

for _ in range(5):
    p = p - lr * (2 * p)          # step opposite the gradient
    path.append(p.copy())
    losses.append(float(p @ p))

path = np.array(path)
```

### Step 2 — Plot the path and the falling loss

The left panel traces the point's path across the plane, with a green star marking the true minimum at `(0, 0)`. The right panel plots `f` at each step — it should fall steadily, confirming that following the negative gradient really does walk downhill.

```python
fig, (ax1, ax2) = plt.subplots(1, 2)

# left — the descent path across the plane
ax1.scatter(path[:, 0], path[:, 1], color="C0")
ax1.plot(path[:, 0], path[:, 1], color="orange")
ax1.scatter([0], [0], color="green", marker="*", s=120)
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.set_title("Descent toward min (0,0)")

# right — the loss falling each step
ax2.plot(range(len(losses)), losses, color="C0")
ax2.set_xlabel("step")
ax2.set_ylabel("f value")
ax2.set_title("Loss going down each step")

plt.show()
```

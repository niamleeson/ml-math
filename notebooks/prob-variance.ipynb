# Variance & standard deviation

_Probability & Statistics_

**The mean tells you the center. Variance tells you the spread.**

Two things can have the same average but feel totally different.

     One is steady; the other swings wildly. The variance measures that swing.

## Reference implementation — NumPy + SciPy

### Step 1 — Variance straight from the definition

A fair die lands on a face `x` in `1..6`, each with probability `1/6`. The mean `mu` is the probability-weighted average of the faces. Variance is then `E[(X - mu)^2]` — the average *squared* distance from the mean — and the standard deviation is just its square root, back in the original units.

```python
import numpy as np

# The six die faces and their (uniform) probabilities.
x = np.arange(1, 7)
p = np.full(6, 1 / 6)

# Mean: probability-weighted average of the faces.
mu = np.sum(x * p)

# Variance: probability-weighted average of squared deviations, E[(X-mu)^2].
var = np.sum((x - mu) ** 2 * p)
std = var ** 0.5

print("Var(X) formula =", round(var, 4), " std =", round(std, 4))
```

### Step 2 — Cross-check against SciPy

SciPy ships a `randint(1, 7)` distribution — the discrete uniform on `1..6`, exactly our die. Its built-in `.var()` and `.std()` should reproduce the numbers we computed by hand, confirming we applied the definition correctly.

```python
from scipy.stats import randint

# scipy's discrete uniform on 1..6 (the upper bound is exclusive).
d = randint(1, 7)

print("scipy  Var =", round(d.var(), 4), " std =", round(d.std(), 4))
```

### Step 3 — Confirm it from samples

Variance is also what you'd *measure* by rolling the die many times. We draw a million rolls and take their sample variance with `ddof=0`, which gives the **population** variance (dividing by N, matching the formula above). With this many samples the estimate lands right on the theoretical value.

```python
# Verify from samples (ddof=0 = population variance).
rng = np.random.default_rng(0)
s = rng.integers(1, 7, size=1_000_000)

sample_var = s.var()
sample_std = s.std()

print("sample Var =", round(sample_var, 4), " std =", round(sample_std, 4))
```

## Visualize the data & results

_How far do die faces sit from the mean, and what is the variance?_

### Step 1 — Compute each face's squared deviation

Variance is the *average* of the squared deviations `(x - mu)^2`. Before averaging, let's look at the individual terms: how far each die face sits from the mean of `3.5`, squared. The faces nearest `3.5` contribute little; the extreme faces `1` and `6` contribute the most.

```python
import numpy as np

# The faces and their uniform probabilities, as before.
x = np.arange(1, 7)
p = np.full(6, 1 / 6)
mu = np.sum(x * p)

# Squared deviation of each face from the mean 3.5.
sq_dev = (x - mu) ** 2
```

### Step 2 — Plot the squared deviations

A bar per face shows its squared distance from the mean. The variance is simply the average height of these bars — a symmetric U-shape, smallest in the middle and largest at the edges.

```python
import matplotlib.pyplot as plt

face_labels = [str(v) for v in x]

plt.bar(face_labels, sq_dev, color='#c89bff')
plt.title('Squared deviations (x-3.5)^2 per face; Var(X) is their average')
plt.xlabel('face')
plt.ylabel('(x - mu)^2')
plt.show()
```

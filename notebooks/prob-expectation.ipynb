# Expectation (the mean)

_Probability & Statistics_

**The long-run average value. Weight each outcome by its chance, then add.**

The expectation is the average value you'd get if you repeated the experiment forever.

     It is a weighted average: each value counts in proportion to how likely it is.

---

This notebook builds the expectation one step at a time — computing `E[X]` for a fair die from its definition, confirming it with scipy, watching a long run of rolls converge to it, and finally drawing the picture.
```python
# Setup — numpy / pandas / scikit-learn / matplotlib ship with Colab.
import numpy as np
import matplotlib.pyplot as plt
```

## Reference implementation — NumPy + SciPy

### Step 1 — Expectation from the definition

For a fair die the outcomes are `1..6`, each with probability `1/6`. The expectation is the probability-weighted sum `E[X] = Σ x·p(x)`. With equal weights this is just the plain average of `1..6`, which comes out to `3.5` — a value the die itself can never land on.

```python
# Fair die: values 1..6, each probability 1/6.
x = np.arange(1, 7)
p = np.full(6, 1 / 6)

EX = np.sum(x * p)                  # E[X] = sum x * p(x)
print("E[X] formula =", round(EX, 4))
```

### Step 2 — Confirm with scipy

`scipy.stats.randint(1, 7)` is the discrete uniform distribution on `1..6` (the high endpoint is exclusive). Its built-in `.mean()` should return the same `3.5`, cross-checking our hand computation.

```python
from scipy.stats import randint

# scipy: discrete uniform on 1..6 (high is exclusive).
print("scipy mean   =", round(randint(1, 7).mean(), 4))
```

### Step 3 — Watch the law of large numbers

Expectation is the *long-run* average. We simulate a million die rolls and take their mean; by the law of large numbers it should sit very close to the theoretical `3.5`. Any small gap shrinks as the number of rolls grows.

```python
# Long-run average converges to E[X] (law of large numbers).
rng = np.random.default_rng(0)
samples = rng.integers(1, 7, size=1_000_000)
print("sample mean  =", round(samples.mean(), 4))
```

## Visualize the data & results

_Where does the probability-weighted average of a die land?_

### Step 1 — Rebuild the PMF and its mean

For the picture we recreate the die's probability mass function — six faces, each with probability `1/6` — and recompute `E[X] = Σ x·p(x)`. These are the same numbers as before, set up here so the plot is self-contained.

```python
# Fair die PMF and its mean E[X] = sum x * p(x).
x = np.arange(1, 7)
p = np.full(6, 1 / 6)
EX = np.sum(x * p)
```

### Step 2 — Plot the PMF with the mean marked

Each face gets an equal-height bar, and a vertical line marks the mean. The bars are drawn at index positions `0..5`, so the face value `3.5` corresponds to x-position `EX - 1 = 2.5` on that axis. The line falls between the `3` and `4` bars — the balance point of the distribution.

```python
plt.bar([str(v) for v in x], p, color='#4ea1ff')
plt.axvline(EX - 1, color='#ffb454', linewidth=2)  # bar index for mean 3.5
plt.title('Fair die PMF with the mean E[X] = 3.5 marked')
plt.xlabel('face')
plt.ylabel('probability')
plt.show()
```

     A coin flip doesn't care what the last flip did. Those flips are independent.

```python
# Setup — numpy / pandas / scikit-learn / matplotlib ship with Colab.
import numpy as np
import matplotlib.pyplot as plt
```

## Reference implementation — NumPy + SciPy

The defining test for independence is simple: events A and B are independent exactly when `P(A and B) = P(A) · P(B)`. We'll check this on two simulated coin flips in three steps: (1) simulate the flips, (2) estimate the individual probabilities, (3) compare the product against the actual joint probability.

### Step 1 — Simulate a million pairs of coin flips

We use a seeded random generator so the result is reproducible, then draw a million values for the first flip and a million for the second. Each value is 0 or 1, with `1` meaning heads. Because the two arrays are drawn separately, the second flip carries no information about the first — exactly the situation independence describes.

```python
rng = np.random.default_rng(0)
n = 1_000_000

flip1 = rng.integers(0, 2, size=n)  # 1 = heads
flip2 = rng.integers(0, 2, size=n)
```

### Step 2 — Estimate the individual probabilities

Let event A be "first flip is heads" and event B be "second flip is heads". We estimate each probability as the fraction of trials where it happened. For a fair coin both should come out close to 0.5.

```python
pA = np.mean(flip1 == 1)  # P(first flip heads)
pB = np.mean(flip2 == 1)  # P(second flip heads)

print("P(A) =", round(pA, 4), " P(B) =", round(pB, 4))
```

### Step 3 — Compare the product with the joint probability

Now the actual test. We estimate `P(A and B)` directly as the fraction of trials where **both** flips were heads, and compare it to the product `P(A) · P(B)`. If the two are close (within a small tolerance), the events pass the independence test — which they should, since the flips were drawn independently.

```python
pAB = np.mean((flip1 == 1) & (flip2 == 1))
product = pA * pB

print("P(A)*P(B) =", round(product, 4), " P(A and B) =", round(pAB, 4))

# Independent exactly when these two match.
print("independent?", abs(product - pAB) < 0.005)
```

## Visualize the data & results

_Are two coin flips independent? Compare P(A)P(B) with P(A and B)._

### Step 1 — Gather the four probabilities to compare

We use the exact theoretical values for two fair coin flips: `P(A) = P(B) = 0.5`. The independence test compares the **product** `P(A)·P(B) = 0.25` against the **true joint** `P(A and B) = 0.25`. Collecting all four side by side sets up a direct visual comparison.

```python
# Theoretical fair-coin probabilities for two flips.
pA = 0.5
pB = 0.5
product = pA * pB
joint = 0.25  # true P(A and B) for two independent fair coins

vals = [pA, pB, product, joint]
```

### Step 2 — Plot the bars

The two rightmost bars are the ones that matter: `P(A)P(B)` and `P(A and B)`. They come out to the same height (0.25), which is the visual signature of independence. If the flips were somehow linked, those two bars would differ.

```python
labels = ['P(A)', 'P(B)', 'P(A)P(B)', 'P(A and B)']
colors = ['#4ea1ff', '#7ee787', '#c89bff', '#ffb454']

plt.bar(labels, vals, color=colors)
plt.title('Independence test: product vs actual joint probability')
plt.ylabel('probability')
plt.show()
```

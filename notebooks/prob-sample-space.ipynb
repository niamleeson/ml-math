## Reference implementation — NumPy + SciPy

We work with the classic example of rolling **two fair dice**. We build it in three steps: (1) enumerate the full sample space, (2) define two events and count their probabilities exactly, (3) confirm those numbers by simulating many rolls.

### Step 1 — Enumerate the sample space

The sample space is the complete list of every outcome that could happen. For two dice, an outcome is an ordered pair `(die1, die2)`, so `itertools.product` over `1..6` twice gives all of them. Because the dice are fair, each of these outcomes is equally likely — that uniformity is what lets us turn counting into probability.

```python
import numpy as np
from itertools import product

# Sample space: all ordered (die1, die2) outcomes -- 36 equally likely.
omega = list(product(range(1, 7), repeat=2))

print("size of sample space:", len(omega))
```

### Step 2 — Define events and count their probabilities

An **event** is any subset of the sample space — the outcomes for which some condition is true. We pick out event A ("the two dice sum to 7") and event B ("both dice are even"). Because every outcome is equally likely, the probability of an event is just its size divided by the size of the sample space.

```python
# Event A = "sum is 7", event B = "both dice even".
A = [o for o in omega if o[0] + o[1] == 7]
B = [o for o in omega if o[0] % 2 == 0 and o[1] % 2 == 0]

pA = len(A) / len(omega)
pB = len(B) / len(omega)

print("P(A=sum7)  =", round(pA, 4), " P(B=both even) =", round(pB, 4))
```

### Step 3 — Verify by simulation

Counting gives the exact answer, but it is reassuring to check it empirically. We roll the two dice a few hundred thousand times and measure the *fraction* of rolls that land in each event. By the law of large numbers these simulated frequencies should sit very close to the exact probabilities from Step 2.

```python
# Verify by simulation.
rng = np.random.default_rng(0)
rolls = rng.integers(1, 7, size=(200000, 2))

simA = np.mean(rolls.sum(axis=1) == 7)
simB = np.mean((rolls[:, 0] % 2 == 0) & (rolls[:, 1] % 2 == 0))

print("sim  P(A) =", round(simA, 4), " P(B) =", round(simB, 4))
```

## Visualize the data & results

_When you roll two fair dice, how likely is each possible sum?_

### Step 1 — Build the distribution of the sum

To draw the picture we again enumerate all 36 outcomes, then collapse each one to its **sum**. Tallying how often each possible sum (2 through 12) appears, divided by 36, gives the probability mass function (PMF) — the probability of each sum.

```python
import numpy as np
import matplotlib.pyplot as plt
from itertools import product

# Enumerate all 36 ordered outcomes and tally each sum.
omega = list(product(range(1, 7), repeat=2))
sums = np.array([a + b for a, b in omega])

vals = np.arange(2, 13)
pmf = np.array([np.mean(sums == s) for s in vals])
```

### Step 2 — Plot the PMF

A bar chart of the PMF shows the familiar triangular shape: 7 is the most likely sum because the most outcome pairs add to it, while 2 and 12 are rarest. We highlight the bar for 7 — the sum from event A above — in green so it stands out.

```python
# Highlight the sum of 7 (event A) in green, the rest in blue.
colors = ['#7ee787' if s == 7 else '#4ea1ff' for s in vals]

plt.bar([str(s) for s in vals], pmf, color=colors)
plt.title('PMF of the sum of two fair dice (36 outcomes)')
plt.xlabel('sum')
plt.ylabel('probability')
plt.show()
```

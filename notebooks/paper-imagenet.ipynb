## Reference implementation — Python

### Step 1 — Write down one WordNet "is-a" branch

ImageNet does not store images as a flat list of labels — it hangs each bucket of images on a node of the **WordNet** noun hierarchy, where every edge means "is-a". To see that structure we hand-type a single root-to-leaf branch, mirroring the mammal subtree in the paper's Figure 1. This is a conceptual sketch of the *tree shape* only — it is not the dataset, not a statistic, and not the paper's figure.

```python
# Conceptual schematic ONLY -- a hand-typed WordNet "is-a" branch.
# This is NOT the ImageNet dataset, NOT the paper's figure, NOT any statistic.
# It just shows the TREE STRUCTURE that ImageNet attaches image buckets to:
# each node is a "synset" (a WordNet concept); an edge means "is-a".

# One root-to-leaf branch, mirroring the paper's Figure 1 (mammal subtree).
branch = [
    "mammal",       # the subtree root (one of the paper's 12 subtrees)
    "placental",
    "carnivore",
    "canine",
    "dog",
    "working dog",
    "husky",        # a leaf: a fine-grained synset
]
```

### Step 2 — Print the branch as an indented is-a tree

Now we render the branch so the nesting is visible. Each item is indented one level deeper than its parent, and every level after the root is prefixed with "is-a" — so each line reads as "this concept is-a kind of the concept above it". The arrow logic is spelled out: depth 0 (the root) gets no "is-a", every deeper level does.

```python
# Print it as an indented "is-a" tree. Each level is a kind of the level above.
print("WordNet is-a branch (conceptual sketch -- not data):\n")

for depth, synset in enumerate(branch):
    indent = "    " * depth                  # one extra level of indent per depth
    arrow = "" if depth == 0 else "is-a "     # the root has no parent, so no "is-a"
    print(f"{indent}{arrow}[{synset}]   <- a synset (image bucket attaches here)")
```

### Step 3 — Spell out what the tree buys you

Finally we print the payoff in words. Every node on this branch is a *synset* — a WordNet concept — and in ImageNet each synset owns its own bucket of images (the paper reports on average over 600 per synset). Because of the is-a edges, a husky image is *also* an instance of dog, canine, ... mammal. That inherited structure is exactly what a flat list of labels would throw away.

```python
print()
print("In ImageNet, EVERY node here is a synset with its own bucket of images")
print("(the paper reports on average >600 images per synset). The husky images")
print("are also, by is-a, instances of dog, canine, ... mammal. That inherited")
print("tree is what a FLAT list of labels would throw away.")

# Expected output (structure only; no dataset is touched):
# [mammal]   <- a synset (image bucket attaches here)
#     is-a [placental]   <- a synset (image bucket attaches here)
#         is-a [carnivore]   <- a synset (image bucket attaches here)
#             is-a [canine]   <- a synset (image bucket attaches here)
#                 is-a [dog]   <- a synset (image bucket attaches here)
#                     is-a [working dog]   <- a synset (image bucket attaches here)
#                         is-a [husky]   <- a synset (image bucket attaches here)
```

### Step 1 — Re-declare the same branch

This panel is read-only — ImageNet is a dataset paper, so there is no model run or curve to plot. We simply re-state the same hand-typed branch as a flat list, independent of the cells above, so this section stands on its own.

```python
# Same conceptual schematic as the CODE panel: one hand-typed WordNet is-a branch.
branch = ["mammal", "placental", "carnivore", "canine", "dog", "working dog", "husky"]
```

### Step 2 — Render the tree as plain text

Instead of drawing a chart we print the same indented is-a tree: indent grows with depth, and every level past the root is labelled "is-a". The output is the bare tree structure — not the ImageNet dataset, not the paper's Figure 1, and no statistics.

```python
# No chart is drawn (this is a read-only dataset paper; there is no curve to plot
# and no run to visualize). We print the tree structure as text instead.
for depth, synset in enumerate(branch):
    indent = "    " * depth                  # deeper concepts indent further
    arrow = "" if depth == 0 else "is-a "     # root has no "is-a" prefix
    print(indent + arrow + "[" + synset + "]")

# [mammal]
#     is-a [placental]
#         is-a [carnivore]
#             is-a [canine]
#                 is-a [dog]
#                     is-a [working dog]
#                         is-a [husky]
# Structure only -- not the ImageNet dataset, not the paper's Figure 1, no statistics.
```

            Name two capabilities you would lose, and tie each to something the paper points out.

<details><summary>Show worked solution</summary>

- Lose fine-vs-coarse grouping: without is-a links you cannot tell that 147 dog breeds all roll up to "dog", or that "dog" rolls up to "mammal". — _The paper highlights dense fine-grained coverage ("147 dog categories") that is only meaningful because the breeds nest under a common parent._

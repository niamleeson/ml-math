## Reference implementation — NumPy + SciPy

We use SciPy's distribution objects to explore both laws. We build it in three steps: (1) the Uniform distribution and its mean, (2) the Exponential distribution and its mean, (3) a simulation that demonstrates the Exponential's memoryless property.

### Step 1 — The Uniform distribution

`Uniform(0, 10)` spreads probability evenly across the interval from 0 to 10 — no value is favoured over any other. SciPy parameterises it by `loc` (the left edge) and `scale` (the width). Its mean sits at the midpoint $(a+b)/2 = 5$, and its CDF at 3 is simply the fraction of the interval below 3.

```python
import numpy as np
from scipy.stats import uniform, expon

# Uniform(0, 10): every value equally likely. Mean = (a+b)/2.
u = uniform(loc=0, scale=10)

print("uniform mean =", round(u.mean(), 4), " P(X<3) =", round(u.cdf(3), 4))
```

### Step 2 — The Exponential distribution

The Exponential models the waiting time until the next event when events arrive at a constant **rate** `lam`. SciPy expects a `scale` argument equal to `1 / lam`, so we convert. With rate `0.5` the mean waiting time is `1 / lam = 2`.

```python
# Exponential with rate lam=0.5 (scipy uses scale = 1/lam). Mean = 1/lam.
lam = 0.5
e = expon(scale=1 / lam)

print("expon mean = 1/lam =", round(e.mean(), 4))
```

### Step 3 — Demonstrate the memoryless property

The Exponential is **memoryless**: having already waited a while tells you nothing about how much longer you must wait, so $P(X > s+t \mid X > s) = P(X > t)$. We test this by sampling two million waiting times, keeping only those that exceeded `s = 2`, and measuring how many of those also exceeded `s + t = 3`. That conditional fraction should match the plain survival probability $P(X > 1)$.

```python
# Memoryless: P(X > s+t | X > s) = P(X > t).
rng = np.random.default_rng(0)
x = rng.exponential(scale=1 / lam, size=2_000_000)

# Given the wait already exceeded 2, how often does it exceed 3 (one more)?
survived_past_2 = x[x > 2.0]
cond = np.mean(survived_past_2 > 2.0 + 1.0)

print("sim P(X>3 | X>2) =", round(cond, 4), " P(X>1) =", round(float(e.sf(1)), 4))
```

## Visualize the data & results

_How do the flat uniform PDF and the decaying exponential PDF differ?_

### Step 1 — Evaluate the two density curves

To compare their shapes we evaluate each probability density function (PDF) on a grid of x-values from 0 to 12. The Uniform PDF is constant at `1/10` inside `[0, 10]` and zero outside; the Exponential PDF starts high and decays toward zero as x grows.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform, expon

# Uniform(0,10) PDF (flat) vs Exponential(rate 0.5) PDF (decay).
xu = np.linspace(0, 12, 200)
u_pdf = uniform(loc=0, scale=10).pdf(xu)

xe = np.linspace(0, 12, 200)
e_pdf = expon(scale=1 / 0.5).pdf(xe)
```

### Step 2 — Plot both PDFs together

Drawing both curves on the same axes makes the contrast obvious: the Uniform is a flat plateau (then a cliff at 10), while the Exponential is a smooth downhill decay. Both still enclose an area of 1, since every PDF integrates to a total probability of one.

```python
plt.plot(xu, u_pdf, color='#7ee787', label='Uniform(0,10)')
plt.plot(xe, e_pdf, color='#4ea1ff', label='Exponential(0.5)')
plt.title('Uniform(0,10) PDF (flat) vs Exponential(rate 0.5) PDF (decay)')
plt.xlabel('x')
plt.ylabel('density')
plt.legend()
plt.show()
```

# Matrix inverse

_Linear Algebra_

**The matrix that undoes another matrix. $A$ maps a shape; $A^{-1}$ maps it back.**

The inverse $A^{-1}$ is the matrix that undoes $A$. Apply $A$, then $A^{-1}$, and you are back where you started: $A^{-1}A = I$.

     Only square matrices can have an inverse, and only if they do not collapse space (that is the determinant from the next lesson).

```python
# Setup — numpy / pandas / scikit-learn / matplotlib ship with Colab.
import numpy as np
import matplotlib.pyplot as plt
```

## Reference implementation — NumPy

We compute a matrix inverse, confirm its defining property, and use it to solve a linear system. We build it in three steps: (1) invert a 2x2 matrix, (2) check that `A A^-1 = I`, (3) solve `A x = b` two ways.

### Step 1 — Invert a 2x2 matrix

`np.linalg.inv(A)` returns the inverse `A^-1`, the matrix that undoes `A`. Only square matrices that don't collapse space (non-zero determinant) have one. We start with a simple invertible 2x2 matrix.

```python
A = np.array([[2., 1.],
              [1., 1.]])
Ainv = np.linalg.inv(A)
print("A^-1 =\n", Ainv)
```

### Step 2 — Check the defining property

By definition, applying `A` then `A^-1` brings you back to where you started, so `A A^-1` must equal the identity. We multiply the two matrices and compare against `np.eye(2)` with `np.allclose`.

```python
# defining property: A A^-1 = I
print("A @ A^-1 = I:", np.allclose(A @ Ainv, np.eye(2)))
```

### Step 3 — Solve A x = b two ways

To solve `A x = b` you can multiply both sides by `A^-1` to get `x = A^-1 b`. That works, but `np.linalg.solve` computes the same `x` directly without forming the inverse, which is faster and numerically more stable. We confirm both routes give the same answer.

```python
# solve A x = b. inverse works, but solve() is faster/stabler
b = np.array([3., 2.])
x_inv = Ainv @ b
x_solve = np.linalg.solve(A, b)

print("x via inverse:", x_inv)
print("x via solve:  ", x_solve)
print("match:", np.allclose(x_inv, x_solve))
```

## Visualize the data & results

_How does A-inverse undo A, returning a vector to exactly where it started?_

### Step 1 — Move a vector with A, then move it back

We take a starting vector `v`, apply `A` to push it to a new spot `Av`, then apply `A^-1` to that result. Because `A^-1` undoes `A`, we land exactly back on the original `v`.

```python
A = np.array([[2., 1.], [1., 1.]])
Ainv = np.linalg.inv(A)          # [[1, -1], [-1, 2]]
v = np.array([3., 5.])
Av = A @ v                       # (11, 8)
back = Ainv @ Av                 # back to (3, 5)
```

### Step 2 — Plot the round trip

We plot the start point `v`, the moved point `Av`, and the two hops between them. The solid line is the forward map `A`; the dashed line is `A^-1` carrying the point back onto `v`, showing the inverse as an exact round trip.

```python
fig, ax = plt.subplots(figsize=(6, 5))
ax.scatter(*v, color='#7ee787', s=80, label='v (start)')
ax.scatter(*Av, color='#4ea1ff', s=80, label='A.v (moved)')
ax.plot([v[0], Av[0]], [v[1], Av[1]], color='#4ea1ff', label='apply A')
ax.plot([Av[0], back[0]], [Av[1], back[1]], color='#c89bff',
        linestyle='--', label='apply A-inverse')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()

plt.tight_layout()
plt.show()
```

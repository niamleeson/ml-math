# Law of Large Numbers

_Probability & Statistics_

**Average enough samples and you home in on the true mean. Why averages are trustworthy.**

One sample is noisy. But average many samples and the noise cancels out.

     The Law of Large Numbers says: as you collect more data, the sample average closes in on the true mean.

```python
# Setup — numpy / pandas / scikit-learn / matplotlib ship with Colab.
import numpy as np
import matplotlib.pyplot as plt
```

## Reference implementation — NumPy + SciPy

### Step 1 — Roll a fair die a million timesThe Law of Large Numbers is about averages settling toward the true mean. Our experiment is a fair six-sided die, whose true mean is (1+2+3+4+5+6)/6 = 3.5. We generate one million rolls up front so we can reuse the *same* stream while growing the sample size.

```python
rng = np.random.default_rng(0)
true_mean = 3.5                     # E[fair die]

# Draw one million rolls in {1,...,6}; integers(1, 7) is upper-exclusive.
rolls = rng.integers(1, 7, size=1_000_000).astype(float)
```

### Step 2 — Watch the running average close in on 3.5Now we take the average of the first n rolls for ever-larger n and compare it to the true mean. With few rolls the average is noisy and the error can be large; as n grows the noise cancels and the error shrinks toward 0 — exactly what the Law of Large Numbers predicts.

```python
# Running average converges to the true mean as n grows.
for n in (10, 100, 1000, 10000, 100000, 1_000_000):
    avg = rolls[:n].mean()
    error = abs(avg - true_mean)
    print("n=%-9d  average=%.4f  error=%.4f" % (n, avg, error))

print("true mean =", true_mean)
```

## Visualize the data & results

_Does the running average of die rolls converge to 3.5?_

### Step 1 — Compute the running average at growing sample sizesTo plot the convergence we pick a range of sample sizes spanning several orders of magnitude and compute the average of the first n rolls at each one. Using the same `rolls` stream means each point is just a longer prefix of the previous one.

```python
# Running average of die rolls at growing sample sizes.
rng = np.random.default_rng(0)
rolls = rng.integers(1, 7, size=1_000_000).astype(float)

ns = np.array([10, 30, 100, 300, 1000, 3000, 10000, 30000,
               100000, 300000, 1_000_000])
avgs = np.array([rolls[:n].mean() for n in ns])
```

### Step 2 — Plot the convergence toward the true meanWe plot the running average against sample size on a log x-axis (so the wide range of n fits) and draw a horizontal line at the true mean 3.5. The curve should wobble for small n and then flatten onto the 3.5 line as the samples accumulate.

```python
plt.plot(ns, avgs, color='#4ea1ff', marker='o', label='running average')
plt.axhline(3.5, color='#ff7b72', label='true mean 3.5')
plt.xscale('log')
plt.title('Law of large numbers: running average of die rolls converges to 3.5')
plt.xlabel('sample size n (log scale labels)')
plt.ylabel('running average')
plt.legend()
plt.show()
```

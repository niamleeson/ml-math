## Reference implementation — NumPy + SciPy

We study **Binomial(n=10, p=0.3)** — the number of successes in 10 independent yes/no trials, each succeeding with probability 0.3. We build it in three steps: (1) read off the exact moments and probabilities from SciPy, (2) simulate the same setup, (3) check the simulation matches.

### Step 1 — Exact moments and probabilities from SciPy

A Binomial is the sum of `n` independent Bernoulli(`p`) trials. SciPy's `binom(n, p)` object hands us everything in closed form: the mean is `n*p`, the variance is `n*p*(1-p)`, the **PMF** gives $P(X = k)$ for an exact count, and the **CDF** gives $P(X \le k)$ for a running total.

We print each one next to the textbook formula so you can see they agree.

```python
from scipy.stats import binom

n = 10
p = 0.3
dist = binom(n, p)

# Mean and variance have closed forms for the Binomial.
print("mean = n*p =", round(dist.mean(), 4), " expected", n * p)
print("var = n*p*(1-p) =", round(dist.var(), 4), " expected", n * p * (1 - p))

# PMF is P(X = k) exactly; CDF is P(X <= k).
print("P(X=3) =", round(dist.pmf(3), 4))
print("P(X<=3) =", round(dist.cdf(3), 4))
```

### Step 2 — Simulate many Binomial draws

To confirm the formulas, we draw half a million Binomial(10, 0.3) outcomes. Each draw is itself the sum of 10 Bernoulli trials, so the sample average should land near the theoretical mean `n*p = 3`, and the fraction of draws equal to exactly 3 should match the PMF from Step 1.

```python
rng = np.random.default_rng(0)

# Each draw is the number of successes in n independent Bernoulli(p) trials.
draws = rng.binomial(n, p, size=500000)

print("sim mean =", round(draws.mean(), 4))
print("sim P(X=3) =", round(np.mean(draws == 3), 4))
```

### Step 3 — Compare simulation to theory

Line the simulated numbers up against the exact values from Step 1. The simulated mean should sit very close to `n*p = 3`, and the simulated $P(X=3)$ should match `dist.pmf(3)` to a couple of decimals — the small leftover gap is just Monte-Carlo noise.

```python
print("sim mean   =", round(draws.mean(), 4), "   theory n*p =", n * p)
print("sim P(X=3) =", round(np.mean(draws == 3), 4), "   theory =", round(dist.pmf(3), 4))
```

## Visualize the data & results

_What does a Binomial(10, 0.3) distribution look like?_

### Step 1 — Build the full PMF

To draw the distribution we need $P(X = k)$ for every possible count `k` from 0 to `n`. We list those counts and ask SciPy for the probability of each — these heights are exactly the bars we will plot.

```python
from scipy.stats import binom

# Binomial(n=10, p=0.3): every possible success count and its probability.
n = 10
p = 0.3
k = np.arange(0, n + 1)
pmf = binom.pmf(k, n, p)
```

### Step 2 — Plot the PMF and mark the mean

We draw one bar per count and highlight the bar at the mean `n*p = 3` in a different color. Notice the distribution leans left (it is skewed) because `p = 0.3` makes small success counts more likely, and the tallest bars cluster around the mean.

```python
# Highlight the mean bar at k = n*p = 3 in orange, the rest in blue.
mean = int(n * p)
colors = ['#ffb454' if j == mean else '#4ea1ff' for j in k]

plt.bar([str(j) for j in k], pmf, color=colors)
plt.title('PMF of Binomial(n=10, p=0.3); mean n*p = 3 marked')
plt.xlabel('successes')
plt.ylabel('probability')
plt.show()
```

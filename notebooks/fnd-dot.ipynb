# The dot product (inner product)

_Foundations: Math you need first_

**Multiply two lists, add it all up, get one number. It measures agreement.**

The dot product takes two vectors and returns a single number.

     Recipe: multiply matching entries, then add everything.

```python
# Setup — numpy / pandas / scikit-learn / matplotlib ship with Colab.
import numpy as np
import matplotlib.pyplot as plt
```

## Reference implementation — NumPy

### Step 1 — Multiply matching entries, then add

A dot product lines up two vectors and pairs off their matching entries. Here `x` describes one house (size, bedrooms, age) and `w` is a price weight for each feature ($ per unit). Multiplying entry-by-entry gives each feature's contribution, and summing them gives a single predicted price. `x @ w` does exactly this and is the same as `np.dot(x, w)`.

```python
x = np.array([1500.0, 3.0, 10.0])         # a house: size, bedrooms, age
w = np.array([200.0, 10000.0, -500.0])    # price weights ($ per unit)

# dot product: multiply matching entries, then add them all up
elementwise = x * w                        # per-feature contributions
dot = x @ w                                # same as np.dot(x, w)

print("elementwise products:", elementwise)
print("w . x  (prediction) :", dot)

# the older the house, the more the negative weight pulls price down
print("matches np.dot      :", np.isclose(dot, np.dot(x, w)))
```

### Step 2 — The same dot product measures agreement

Beyond predicting a number, the dot product tells you how much two directions agree. Dividing the dot product by the two vectors' lengths gives the **cosine similarity**: `+1` means they point the same way, `0` means perpendicular, `-1` means opposite. Here `a` and `b` lean in similar-but-not-identical directions, so the cosine lands between 0 and 1.

```python
a = np.array([3.0, 1.0])
b = np.array([1.0, 3.0])

# cosine similarity = dot product divided by the two lengths
dot_ab = a @ b
len_a = np.linalg.norm(a)
len_b = np.linalg.norm(b)
cos = dot_ab / (len_a * len_b)

print("cosine(a, b)        :", round(float(cos), 4))
```

## Visualize the data & results

_Does the dot product tell you whether two vectors agree?_

### Step 1 — Dot a fixed vector with three directions

To see the dot product as an agreement score, we fix a base vector `[1, 0]` and dot it with three other directions: one that agrees, one that is perpendicular, and one that opposes. The sign of each result is the giveaway — positive for agreement, zero for perpendicular, negative for opposition.

```python
# dot [1,0] with three directions: agree, perpendicular, oppose
base = np.array([1.0, 0.0])
dirs = np.array([[3.0, 0.0], [0.0, 5.0], [-4.0, 0.0]])
dots = dirs @ base                       # [3, 0, -4]

print("dot products:", dots)
```

### Step 2 — Plot the agreement scores and two vectors

Now we draw the picture. The left bar chart colours each direction by its dot product — green for agreement, gray for perpendicular, red for opposition. The right panel re-computes the cosine to show two vectors `a` and `b` that partly agree, with their cosine similarity in the title.

```python
a = np.array([3.0, 1.0])
b = np.array([1.0, 3.0])
cos = (a @ b) / (np.linalg.norm(a) * np.linalg.norm(b))  # 0.6

fig, (ax1, ax2) = plt.subplots(1, 2)

# left — dot product of [1,0] with each direction
labels = ["same [3,0]", "perp [0,5]", "opp [-4,0]"]
ax1.bar(labels, dots, color=["green", "gray", "red"])
ax1.set_title("Dot product of [1,0] with each direction")

# right — the two vectors a and b
ax2.quiver([0, 0], [0, 0], [a[0], b[0]], [a[1], b[1]],
           angles="xy", scale_units="xy", scale=1, color=["C0", "purple"])
ax2.set_xlim(-1, 4)
ax2.set_ylim(-1, 4)
ax2.set_xlabel("x1")
ax2.set_ylabel("x2")
ax2.set_title("a, b  cosine = " + str(round(float(cos), 2)))

plt.show()
```

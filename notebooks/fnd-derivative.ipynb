**How fast does the output change when you nudge the input? That's the derivative.**

A derivative is a slope: the rate of change of a function.

     It answers: "if I increase the input a tiny bit, how much does the output move, and in which direction?"

## Reference implementation — NumPy

### Step 1 — Define the function and its exact slope

We start with a simple bowl-shaped function $f(x) = x^2$. Its derivative — the slope at any point — is known exactly from calculus: the slope of $x^2$ is $2x$. We write both as plain Python functions so we can compare them.

```python
import numpy as np

def f(x):                 # a U-shaped bowl
    return x ** 2

def f_prime(x):           # exact derivative: slope of x^2 is 2x
    return 2 * x
```

### Step 2 — Approximate the slope numerically

A derivative is the limit of $(f(x+h) - f(x))/h$ as the nudge $h$ shrinks to zero. We can't take a true limit on a computer, so we pick a tiny $h$ and compute the ratio directly. This **finite-difference** estimate should land very close to the exact $2x$ at each test point.

```python
# finite-difference approximation of the limit (f(x+h)-f(x))/h
h = 1e-6
xs = np.array([-2.0, 0.0, 3.0])

approx = (f(xs + h) - f(xs)) / h
exact = f_prime(xs)
```

### Step 3 — Compare the two and check the minimum

We print the numeric estimate next to the exact slope for each point — they should match closely. Notice the slope is exactly $0$ at $x=0$: that is the bottom of the bowl, a minimum, where the function momentarily stops changing.

```python
for xi, ap, ex in zip(xs, approx, exact):
    print("x=%5.1f  approx slope=%8.4f  exact 2x=%6.1f" % (xi, ap, ex))

# slope is 0 at the bottom of the bowl (x=0) -> a minimum
print("close to exact?", np.allclose(approx, exact, atol=1e-3))
```

## Visualize the data & results

_Why is the slope of x squared equal to 2x?_

### Step 1 — Plot the curve and a tangent line

To *see* what a derivative means, we draw the bowl $f(x) = x^2$ and overlay the tangent line at $x=1.5$. The tangent has slope $2 \times 1.5 = 3$, and it just grazes the curve at that point — its steepness is exactly the derivative there.

```python
import numpy as np
import matplotlib.pyplot as plt

xs = np.linspace(-3, 3, 13)
f = xs ** 2                       # the bowl

# tangent at x = 1.5 has slope 2*1.5 = 3
x0 = 1.5
f_at_x0 = f[xs == x0][0]          # height of the curve at x0
tangent = f_at_x0 + 3 * (xs - x0) # straight line through (x0, f(x0)) with slope 3
```

### Step 2 — Show the exact slope at three points

On the right we compute the exact derivative $2x$ at three sample points and draw them as bars. The signs tell the story: negative slope on the left of the bowl, zero at the bottom, positive on the right.

```python
pts = np.array([-2.0, 0.0, 3.0])
slopes = 2 * pts                  # exact derivative 2x: [-4, 0, 6]

fig, (ax1, ax2) = plt.subplots(1, 2)

# Left — the bowl and its tangent at x=1.5
ax1.plot(xs, f, color="C0", label="f(x) = x^2")
ax1.plot(xs, tangent, color="orange", label="tangent at x=1.5")
ax1.set_xlabel("x")
ax1.set_ylabel("f(x)")
ax1.set_ylim(-4, 9)
ax1.legend()
ax1.set_title("x^2 and its tangent")

# Right — the exact slope 2x at three points
ax2.bar(["x=-2", "x=0", "x=3"], slopes, color=["red", "gray", "green"])
ax2.set_title("Exact derivative 2x")

plt.show()
```

# Cosine similarity

_Deep Learning_

**Measure how alike two vectors are by the angle between them, not their length.**

Cosine similarity measures how similar two vectors point, ignoring how long they are.

     It is the cosine of the angle between them. Pointing the same way gives 1. At a right angle gives 0. Opposite gives -1.

## Reference implementation — NumPy

### Step 1 — Define the cosine similarity

Cosine similarity is the dot product of two vectors divided by the product of their lengths (their L2 norms). Dividing by the norms cancels out how *long* each vector is, leaving only how they are *oriented*. The result always lands between -1 (opposite directions) and +1 (same direction).

```python
def cosine(a, b):
    dot = a @ b                       # dot product measures aligned overlap
    norm_a = np.linalg.norm(a)        # length of a
    norm_b = np.linalg.norm(b)        # length of b
    return dot / (norm_a * norm_b)    # divide out the lengths, keep only direction
```

### Step 2 — Compare a few vectors by direction

We build three vectors. `b` is `a` scaled up, so it points the exact same way; `c` points off in a different direction. Cosine similarity should report `a` and `b` as a perfect 1.0 because they share a direction, while `a` and `c` score lower.

```python
a = np.array([1.0, 2.0, 3.0])
b = np.array([2.0, 4.0, 6.0])    # same direction as a (just scaled)
c = np.array([-1.0, 0.0, 0.0])

print("a vs b (parallel):", round(cosine(a, b), 4))    # 1.0
print("a vs c (apart)   :", round(cosine(a, c), 4))
```

### Step 3 — Confirm length doesn't matter

To drive the point home, we compare `a` against `100 * a` — the same vector stretched a hundred times longer. Because cosine similarity divides out the lengths, the score is still exactly 1.0: only the shared direction counts.

```python
# length does not matter, only direction:
print("a vs 100*a       :", round(cosine(a, 100 * a), 4))   # still 1.0
```

## Visualize the data & results

_Using real word vectors, how does cosine similarity score related vs unrelated word pairs?_

### Step 1 — Set up toy word vectors

Real language models represent each word as a vector, where related words point in similar directions. Here we hand-pick a small dictionary of four-number vectors — royalty, cities, and animals — so we can test whether cosine similarity recovers the relationships we'd expect.

```python
vec = {
    "king":   [0.9, 0.95, 0.1, 0.8],
    "queen":  [0.88, 0.1, 0.12, 0.82],
    "paris":  [-0.8, 0.4, 0.9, -0.6],
    "london": [-0.78, 0.45, 0.88, -0.5],
    "france": [-0.85, 0.42, 0.95, -0.55],
    "cat":    [0.1, -0.85, -0.7, 0.3],
    "dog":    [0.12, -0.82, -0.72, 0.28],
}

# turn each plain list into a NumPy array of floats
vec = {word: np.array(values, float) for word, values in vec.items()}

def cosine(a, b):
    dot = a @ b
    norm_a = np.linalg.norm(a)
    norm_b = np.linalg.norm(b)
    return dot / (norm_a * norm_b)
```

### Step 2 — Score related vs unrelated pairs

We pick five word pairs: three that should be similar (king/queen, paris/london, cat/dog) and two that should not (king/cat, king/france). We compute the cosine similarity of each pair and colour them green for the related pairs and red for the unrelated ones, so the plot will make the split obvious.

```python
pairs = [
    ("king", "queen"),
    ("paris", "london"),
    ("cat", "dog"),
    ("king", "cat"),
    ("king", "france"),
]

labels = [a + " vs " + b for a, b in pairs]
values = [cosine(vec[a], vec[b]) for a, b in pairs]
colors = ["#7ee787", "#7ee787", "#7ee787", "#ff7b72", "#ff7b72"]
```

### Step 3 — Plot the similarities

A bar chart of the five scores. The related pairs should rise toward 1.0 while the unrelated pairs sit much lower — visual confirmation that cosine similarity captures meaning, not coincidence.

```python
plt.bar(labels, values, color=colors)
plt.title("Cosine similarity between real word-vector pairs")
plt.axhline(0, color="gray", linewidth=0.8)
plt.xticks(rotation=20)
plt.show()
```

```python
# Setup — numpy / pandas / scikit-learn / matplotlib ship with Colab.
import numpy as np
import matplotlib.pyplot as plt
```

## Reference implementation — NumPy

### Step 1 — Build one vector and read its shape

A vector is simply an ordered list of numbers. Here we describe a single house with three measurements — size in square feet, number of bedrooms, and age in years — and store them in a NumPy array. The array's `dtype` tells us the kind of numbers it holds, and its `shape` of `(n,)` tells us it is a 1-D list of length `n`.

```python
# one house as a vector: [size_sqft, bedrooms, age_years]
x = np.array([1500.0, 3.0, 10.0])

print("vector x   :", x)
print("dtype      :", x.dtype)     # numbers it holds
print("shape      :", x.shape)     # (n,) -> a 1-D array of length n
print("n (length) :", x.shape[0])
```

### Step 2 — Read individual entries by position

Each number in the vector lives at a position, and NumPy counts those positions starting from `0`. So `x[0]` is the first entry (size), `x[1]` the second (bedrooms), and `x[2]` the third (age). Indexing is how we pull one feature out of the list when we need it.

```python
# read entries by position (NumPy indexes from 0)
print("x[0] size     :", x[0])
print("x[1] bedrooms :", x[1])
print("x[2] age      :", x[2])
```

### Step 3 — Stack vectors into a dataset

A dataset is nothing more than many vectors of the same shape. Here we add a second house described by the very same three features. Every row you collect is one more vector, and together they form the data a model learns from.

```python
# a dataset is many such vectors; here is a second house
y = np.array([900.0, 2.0, 25.0])
print("second house :", y)
```

## Visualize the data & results

_What does the house vector [1500, 3, 10] look like, and how long is it?_

### Step 1 — Set up the figure and the vectors to draw

We will show two pictures side by side, so we ask Matplotlib for a figure with two axes. We also keep the full house vector for the bar chart, plus a small 2-D vector `[3, 2]` that is easy to draw as an arrow on a flat plane (we can only sketch arrows in two dimensions).

```python
# the house vector and a small 2-D vector to draw as an arrow
house = np.array([1500.0, 3.0, 10.0])
x = np.array([3.0, 2.0])

fig, (ax1, ax2) = plt.subplots(1, 2)
```

### Step 2 — Draw the 2-D vector as an arrow

On the left axis we draw `x = [3, 2]` as an arrow starting at the origin. The dashed orange line shows how far it reaches along x1, and the dashed green line shows how far it climbs along x2. Together those two legs are exactly the two numbers stored in the vector.

```python
# left: x drawn as an arrow from the origin with its legs
ax1.quiver(0, 0, x[0], x[1], angles="xy", scale_units="xy", scale=1, color="C0")
ax1.plot([0, x[0]], [0, 0], "--", color="orange")
ax1.plot([x[0], x[0]], [0, x[1]], "--", color="green")
ax1.set_xlim(-1, 4)
ax1.set_ylim(-1, 3)
ax1.set_xlabel("x1")
ax1.set_ylabel("x2")
ax1.set_title("Vector x = [3, 2] as an arrow")
```

### Step 3 — Draw the house vector's entries as bars

The house vector has three entries, so a flat arrow no longer fits. Instead we show each entry as its own bar on the right axis. This view makes it clear that a vector is just a labelled list of numbers, however many of them there are. Calling `plt.show()` renders both panels.

```python
# right: the three entries of the house vector as bars
labels = ["x0 size", "x1 bedrooms", "x2 age"]
ax2.bar(labels, house, color=["C0", "green", "orange"])
ax2.set_title("Entries of x = [1500, 3, 10]")
plt.show()
```

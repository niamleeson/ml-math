## Reference implementation — NumPy + SciPy

We'll look at the two distributions side by side: the **Geometric** counts trials until the first success, and the **Poisson** counts rare events in a fixed window. We build it in three steps: (1) the Geometric, (2) the Poisson, (3) a simulation check that the formulas match reality.

### Step 1 — The Geometric distribution

`Geometric(p)` models the number of trials up to and including the **first success**, when each trial succeeds independently with probability `p`. Its mean is `1/p` — if a success happens 25% of the time, you wait 4 trials on average. We use SciPy's `geom(p)` to read off the mean and the probability that the first success lands exactly on trial 3.

```python
from scipy.stats import geom

# Geometric(p): trials until first success. Mean = 1/p.
p = 0.25
g = geom(p)

geom_mean = g.mean()
print("geom mean = 1/p =", round(geom_mean, 4), " expected", 1 / p)
print("geom P(first success on trial 3) =", round(g.pmf(3), 4))
```

### Step 2 — The Poisson distribution

`Poisson(lam)` models the **count of events** in a fixed window when events arrive at an average rate `lam`. A defining feature is that its mean and variance are **both equal to** `lam`. We use SciPy's `poisson(lam)` to confirm that, and to read the probability of seeing exactly 4 events.

```python
from scipy.stats import poisson

# Poisson(lam): count of events in a window. Mean = var = lam.
lam = 4.0
po = poisson(lam)

poisson_mean = po.mean()
poisson_var = po.var()
print("poisson mean =", round(poisson_mean, 4), " var =", round(poisson_var, 4))
print("poisson P(X=4) =", round(po.pmf(4), 4))
```

### Step 3 — Check the means by simulation

Formulas are easy to misremember, so we sanity-check them. We draw half a million samples from each distribution with NumPy's random generator and average them. The simulated means should land right on `1/p = 4` for the Geometric and `lam = 4` for the Poisson.

```python
# Simulate both means with a seeded generator (reproducible).
rng = np.random.default_rng(0)

geom_samples = rng.geometric(p, 500000)
poisson_samples = rng.poisson(lam, 500000)

print("sim geom mean   =", round(geom_samples.mean(), 4))
print("sim poisson mean=", round(poisson_samples.mean(), 4))
```

## Visualize the data & results

_How do the geometric and Poisson PMFs compare in shape?_

### Step 1 — Build the two PMFs

We evaluate each distribution's probability mass function over a range of outcomes. The Geometric starts at trial 1 (you need at least one trial), while the Poisson starts at 0 (a window can contain zero events). `geom.pmf` and `poisson.pmf` give the probability of each individual outcome.

```python
from scipy.stats import geom, poisson

# Geometric(0.25): trial of first success. Starts at 1.
gk = np.arange(1, 16)
g_pmf = geom.pmf(gk, 0.25)

# Poisson(4): events per window. Starts at 0.
pk = np.arange(0, 16)
p_pmf = poisson.pmf(pk, 4.0)
```

### Step 2 — Plot them side by side

The Geometric PMF decays steadily — the most likely outcome is a success on the very first trial, with each later trial less likely. The Poisson PMF is humped, peaking near its mean of 4. Plotting both makes the difference in shape obvious.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(11, 4))

# Left — Geometric PMF
ax1.bar([str(v) for v in gk], g_pmf, color='#7ee787')
ax1.set_title('Geometric(p=0.25): trial of first success (mean 1/p = 4)')

# Right — Poisson PMF
ax2.bar([str(v) for v in pk], p_pmf, color='#4ea1ff')
ax2.set_title('Poisson(lambda=4): events per window (mean = var = 4)')

plt.show()
```

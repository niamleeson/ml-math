## Reference implementation — NumPy

### Step 1 — Multiply two matrices with `@`

We start with two 2x2 matrices `A` and `B`. NumPy's `@` operator is matrix multiplication: entry `(i, j)` of the product is row `i` of `A` dotted with column `j` of `B`. Let NumPy do the whole product in one call so we have a reference answer to compare against.

```python
import numpy as np

A = np.array([[1, 2],
              [3, 4]])
B = np.array([[5, 6],
              [7, 8]])

# @ is matrix multiply: (AB)_ij = sum_k A_ik B_kj
AB = A @ B
print("A @ B =\n", AB)
```

### Step 2 — Rebuild the product entry by entry

To see what `@` actually does, we reconstruct the product one entry at a time. For each output position `(i, j)` we take row `i` of `A` and column `j` of `B` and dot them together. Stacking those dot products into a matrix should reproduce `AB` exactly.

```python
# rebuild it entry by entry from dot products to show what @ does
m = A.shape[0]          # number of rows in the result
p = B.shape[1]          # number of columns in the result

manual = np.array([[A[i, :] @ B[:, j] for j in range(p)] for i in range(m)])
print("manual =\n", manual)

print("match:", np.allclose(AB, manual))
```

### Step 3 — Order matters: AB is not BA

Matrix multiplication is **not commutative**. Swapping the operands generally changes the result, because the rows and columns being dotted are different. We compute `B @ A` and confirm it differs from `A @ B`.

```python
# order matters: AB != BA in general
print("AB == BA?", np.allclose(A @ B, B @ A))
```

## Visualize the data & results

_Where does each entry of A @ B come from, and why is AB different from BA?_

### Step 1 — Compute both orderings

To visualise that order matters, we compute both products with the same `A` and `B`. `AB` and `BA` come out as different 2x2 matrices, which we will display side by side.

```python
import numpy as np
import matplotlib.pyplot as plt

A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
AB = A @ B           # 19,22 / 43,50
BA = B @ A           # 23,34 / 31,46
```

### Step 2 — Show the two products as heatmaps

We draw `AB` and `BA` as colour grids with the numeric value printed in each cell. Seeing the two grids next to each other makes the non-commutativity concrete: same inputs, different entries.

```python
fig, axes = plt.subplots(1, 2, figsize=(8, 4))

for ax, M, title in zip(axes, [AB, BA], ['A @ B', 'B @ A']):
    ax.imshow(M, cmap='Blues')
    for i in range(2):
        for j in range(2):
            ax.text(j, i, str(M[i, j]), ha='center', va='center')
    ax.set_title(title)
    ax.set_xticks([0, 1])
    ax.set_yticks([0, 1])

plt.tight_layout()
plt.show()
```

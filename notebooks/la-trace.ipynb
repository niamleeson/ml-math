# Trace

_Linear Algebra_

**Add up the diagonal. It equals the sum of the eigenvalues.**

The trace is the sum of the diagonal entries of a square matrix.

     It looks trivial, but it has two deep properties: you can swap the order in a product, $\operatorname{tr}(AB)=\operatorname{tr}(BA)$, and it always equals the sum of the matrix's eigenvalues.

## Reference implementation — NumPy

### Step 1 — Add up the diagonal

The trace is just the sum of the entries running down the main diagonal of a square matrix. For our `A` that is `2 + 3 + 6 = 11`, which `np.trace` computes for us.

```python
import numpy as np

A = np.array([[2., 1., 0.],
              [4., 3., 5.],
              [1., 0., 6.]])

# The trace sums the diagonal entries: 2 + 3 + 6 = 11.
print("trace(A) =", np.trace(A))
```

### Step 2 — Check that the trace equals the sum of the eigenvalues

A deeper fact: the trace always equals the sum of a matrix's eigenvalues. We pull the eigenvalues out with `np.linalg.eigvals`, take the real part of their sum (eigenvalues can be complex in general), and confirm it matches the diagonal sum from Step 1.

```python
# The trace also equals the sum of the eigenvalues.
eigs = np.linalg.eigvals(A)
eig_sum = eigs.sum().real
print("sum of eigenvalues =", round(eig_sum, 6))
print("trace == sum(eig):", np.isclose(np.trace(A), eig_sum))
```

### Step 3 — Verify the cyclic property tr(AB) = tr(BA)

The trace is invariant when you swap the order of a product: `tr(AB) = tr(BA)`, even though `AB` and `BA` are different matrices. We build a second matrix `B` and check both products land on the same trace.

```python
# Cyclic property: tr(AB) == tr(BA), even though AB != BA.
B = np.array([[1., 0., 2.],
              [3., 1., 0.],
              [0., 4., 1.]])

trace_AB = np.trace(A @ B)
trace_BA = np.trace(B @ A)
print("tr(AB) == tr(BA):", np.isclose(trace_AB, trace_BA))
```

## Visualize the data & results

_Why does the trace (sum of the diagonal) equal the sum of the eigenvalues?_

### Step 1 — Collect the diagonal entries and the eigenvalues

To show the trace-equals-eigenvalue-sum fact visually, we grab two sets of numbers: the diagonal entries of `A` (which sum to the trace, 11) and the eigenvalues sorted small-to-large (whose real parts sum to the same 11).

```python
import numpy as np
import matplotlib.pyplot as plt

A = np.array([[2., 1., 0.],
              [4., 3., 5.],
              [1., 0., 6.]])

diag = np.diag(A)                          # 2, 3, 6 -> trace 11
eigs = np.sort(np.linalg.eigvals(A).real)  # ~0.68, 3.88, 6.44
```

### Step 2 — Plot both sums side by side

Left bar chart: the three diagonal entries, whose heights add up to the trace. Right bar chart: the three eigenvalues, whose heights add up to the very same total. We annotate each bar with its value so the two equal sums are easy to compare.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(9, 4))

# Left — the diagonal entries, summing to the trace.
ax1.bar(['A[0,0]=2', 'A[1,1]=3', 'A[2,2]=6'], diag, color='#4ea1ff')
ax1.set_title('diagonal sums to trace = ' + str(int(diag.sum())))

# Right — the eigenvalues, summing to the same number.
ax2.bar(['lambda 1', 'lambda 2', 'lambda 3'], eigs, color='#c89bff')
ax2.set_title('eigenvalues sum to ' + str(round(eigs.sum())))

# Label every bar with its value.
for ax, vals in [(ax1, diag), (ax2, eigs)]:
    for i, v in enumerate(vals):
        ax.text(i, v, str(round(v, 2)), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

## Reference implementation — NumPy + SciPy

### Step 1 — Compute the PMF of a Binomial random variable

Let `X` be the number of heads in 3 fair coin flips. `X` is a **random variable**: it maps each outcome to a number in `{0, 1, 2, 3}`. Its **PMF** `P(X = k)` lists how likely each value is — here a `Binomial(n=3, p=0.5)`. Because the PMF is a full probability distribution, the values must sum to 1.

```python
import numpy as np
from scipy.stats import binom

n = 3
p = 0.5
k = np.arange(0, n + 1)
pmf = binom.pmf(k, n, p)            # P(X = k)

print("PMF P(X=k):", np.round(pmf, 4))
print("sums to:", round(pmf.sum(), 6))
```

### Step 2 — Confirm the PMF by simulating coin flips

To check the formula empirically we simulate 500,000 trials of 3 flips each, count the heads per trial, and measure how often each head-count occurs. Those empirical frequencies should closely match the exact PMF from Step 1.

```python
# Verify by simulation: count heads in many 3-flip trials.
rng = np.random.default_rng(0)
flips = rng.integers(0, 2, size=(500000, n))
heads = flips.sum(axis=1)

sim = np.array([np.mean(heads == j) for j in k])
print("sim P(X=k):", np.round(sim, 4))
```

## Visualize the data & results

_What is the distribution of heads in 3 fair flips?_

### Step 3 — Recompute the PMF for plotting

We rebuild the same `Binomial(3, 0.5)` PMF so the visualization cell stands on its own. `k` holds the possible head-counts and `pmf` their probabilities.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# X = heads in 3 fair flips ~ Binomial(3, 0.5).
n = 3
p = 0.5
k = np.arange(0, n + 1)
pmf = binom.pmf(k, n, p)
```

### Step 4 — Plot the PMF as a bar chart

Each bar shows `P(X = k)` for one head-count. The chart is symmetric and peaks at 1 and 2 heads — the most likely outcomes — while 0 and 3 heads are rarer.

```python
labels = [str(j) for j in k]
plt.bar(labels, pmf, color='#4ea1ff')
plt.title('PMF of X = heads in 3 flips  ~ Binomial(3, 0.5)')
plt.xlabel('heads')
plt.ylabel('probability')
plt.show()
```

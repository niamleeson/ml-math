# The chain rule

_Foundations: Math you need first_

**Derivative of a function inside a function: multiply the slopes. This IS backprop (backpropagation).**

Often a value passes through several steps: $x \to$ step 1 $\to$ step 2 $\to$ output.

     The chain rule says: to get the overall slope, multiply the slope of each step together.

## Reference implementation — NumPy

### Step 1 — Build a composition of two functions

The chain rule is about functions nested inside functions. Here the input $x$ first goes through an **inner** step $y = 3x$, and that result feeds an **outer** step $z = y^2$. Composing them gives $z = (3x)^2$.

```python
import numpy as np

# composition: y = 3x  (inner),  z = y^2  (outer)
def inner(x):
    return 3 * x

def outer(y):
    return y ** 2

def z(x):
    return outer(inner(x))
```

### Step 2 — Apply the chain rule

The chain rule says: multiply the slopes of each step. The outer slope is $dz/dy = 2y$ and the inner slope is $dy/dx = 3$. Multiplying gives $dz/dx = 2y \cdot 3 = 6(3x) = 18x$. We code this directly: compute the inner value first, then multiply the two slopes.

```python
# chain rule: dz/dx = dz/dy * dy/dx = (2y) * 3 = 6*(3x) = 18x
def dz_dx(x):
    y = inner(x)
    outer_slope = 2 * y        # dz/dy
    inner_slope = 3            # dy/dx
    return outer_slope * inner_slope
```

### Step 3 — Check it against a numeric slope

To confirm the chain rule gave the right answer, we estimate the true slope numerically with a tiny nudge $h$, then compare it to both our chain-rule formula and the closed form $18x$. All three should agree at every test point.

```python
xs = np.array([-1.0, 0.5, 1.0])
h = 1e-6

approx = (z(xs + h) - z(xs)) / h     # numeric check of the limit
exact = dz_dx(xs)

for xi, ap, ex in zip(xs, approx, exact):
    print("x=%5.2f  chain-rule=%7.3f  numeric=%7.3f  (18x=%6.2f)" % (xi, ex, ap, 18 * xi))

print("chain rule matches numeric?", np.allclose(approx, exact, atol=1e-3))
```

## Visualize the data & results

_Does the chain rule give the right slope for z = (3x) squared?_

### Step 1 — Evaluate the slope at three points

Because the chain rule collapsed to $dz/dx = 18x$, the slope is itself a simple straight line. We first pick three sample points and compute their slopes, then build the full line $18x$ across the range so we can plot it.

```python
import numpy as np
import matplotlib.pyplot as plt

# z = (3x)^2, chain rule dz/dx = (2y)*3 = 18x
pts = np.array([-1.0, 0.5, 1.0])
slopes = 18 * pts                 # [-18, 9, 18]

line_x = np.linspace(-1, 1, 5)
line_y = 18 * line_x              # the straight line 18x
```

### Step 2 — Plot the slopes and the slope line

On the left we draw the slope at each sample point as a bar (note the negative bar at $x=-1$). On the right we draw the continuous slope function $18x$, confirming the derivative is a straight line through the origin.

```python
fig, (ax1, ax2) = plt.subplots(1, 2)

# Left — slope at three sample points
ax1.bar(["x=-1", "x=0.5", "x=1"], slopes, color=["red", "green", "C0"])
ax1.set_title("Chain rule dz/dx = 18x")

# Right — the slope as a continuous line
ax2.plot(line_x, line_y, color="purple")
ax2.set_xlabel("x")
ax2.set_ylabel("dz/dx")
ax2.set_title("Slope of (3x)^2 is the line 18x")

plt.show()
```

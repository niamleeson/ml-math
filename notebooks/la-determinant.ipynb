# Determinant

_Linear Algebra_

**One number that says how much a matrix scales area. Zero means it collapses space.**

The determinant is a single number squeezed out of a square matrix.

     Its meaning is geometric: $|\det A|$ is the factor by which $A$ scales area (in 2D) or volume (in 3D). A unit square of area 1 becomes a parallelogram of area $|\det A|$.

## Reference implementation — NumPy

### Step 1 — Compute a 2×2 determinant

For a 2×2 matrix the determinant is `ad - bc`. We let NumPy compute `det(A)` and confirm it equals that hand formula. A nonzero determinant (here 5) means `A` is invertible and scales area by a factor of 5.

```python
import numpy as np

A = np.array([[3., 1.],
              [1., 2.]])

det = np.linalg.det(A)
print("det(A) =", round(det, 6))                 # 3*2 - 1*1 = 5
print("matches ad-bc:", np.isclose(det, 3*2 - 1*1))
```

### Step 2 — See what a zero determinant means

Now take a matrix whose second column is just twice its first — the columns are parallel, so the matrix collapses 2D space onto a line. Its determinant is 0, and a determinant of 0 means there is **no inverse**: `np.linalg.inv` raises a `LinAlgError`, which we catch and print.

```python
# singular matrix: parallel columns -> det 0 -> no inverse
S = np.array([[2., 4.],
              [1., 2.]])
print("det(S) =", round(np.linalg.det(S), 6))    # 0

try:
    np.linalg.inv(S)
except np.linalg.LinAlgError as e:
    print("inverse failed (singular):", e)
```

## Visualize the data & results

_Why is det(A) the factor by which A scales area?_

### Step 1 — Transform the unit square

The determinant's geometric meaning: `|det A|` is the area-scaling factor. We start from the unit square (area 1) and apply `A` to each of its corners. Because each corner is a row vector, we multiply by `A.T` so every point is transformed by `A`.

```python
import numpy as np
import matplotlib.pyplot as plt

A = np.array([[3., 1.], [1., 2.]])
det = np.linalg.det(A)                   # 5.0

# corners of the unit square, closed back to the start for plotting
square = np.array([[0, 0], [1, 0], [1, 1], [0, 1], [0, 0]])
image = square @ A.T                      # transform each corner by A
print("area scaling factor |det A| =", round(det))
```

### Step 2 — Plot the square and its image

We draw the original unit square (dashed) and the parallelogram it becomes after `A`. The parallelogram's area equals `|det A|` = 5 — a unit square of area 1 is stretched into a shape five times larger, which is exactly what the determinant measures.

```python
fig, ax = plt.subplots(figsize=(6, 5))

# original unit square
ax.plot(square[:, 0], square[:, 1], color='#9aa7b4',
        linestyle='--', label='unit square (area 1)')

# image parallelogram after applying A
ax.plot(image[:, 0], image[:, 1], color='#4ea1ff',
        label='image parallelogram (area ' + str(round(det)) + ')')
ax.scatter(image[:-1, 0], image[:-1, 1], color='#4ea1ff')

ax.set_title('|det A| = ' + str(round(det)))
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
plt.tight_layout()
plt.show()
```

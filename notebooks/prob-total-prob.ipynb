## Reference implementation — NumPy + SciPy

We use a factory example: two factories make the same part at different defect rates, and we want the overall chance a random part is defective. We build it in three steps: (1) lay out the cases and their probabilities, (2) apply the total probability theorem, (3) verify by simulation.

### Step 1 — Lay out the cases and conditional probabilities

The "cases" are a partition of the world into mutually exclusive, exhaustive events — here, which factory a part came from. `pA` holds the share of parts from each factory (they sum to 1), and `pB_given` holds the conditional defect rate *within* each factory. The defect rate is easy to state inside each case even though the overall rate is not obvious yet.

```python
import numpy as np

pA = np.array([0.6, 0.4])           # P(factory 1), P(factory 2)
pB_given = np.array([0.02, 0.05])   # P(defect | factory)

print("factory shares:", pA, " defect rates:", pB_given)
```

### Step 2 — Apply the total probability theorem

The theorem says $P(B) = \sum_i P(A_i)\,P(B \mid A_i)$: weight each case's conditional probability by how likely that case is, then add. The element-wise product `pA * pB_given` gives each case's contribution, and summing them collapses the cases into the single overall defect probability.

```python
# Total probability theorem: sum_i P(A_i) P(B | A_i).
pB = np.sum(pA * pB_given)

print("P(defect) =", round(pB, 4))
```

### Step 3 — Verify by simulation

To check the formula we simulate a million parts. First we draw which factory each part comes from using the shares `pA`, then we flag it defective with that factory's defect rate. The fraction of defective parts should land very close to the `P(defect)` computed in Step 2.

```python
# Verify by simulation.
rng = np.random.default_rng(0)
n = 1_000_000

factory = rng.choice([0, 1], size=n, p=pA)
defect = rng.random(n) < pB_given[factory]

print("sim P(defect) =", round(np.mean(defect), 4))
```

## Visualize the data & results

_What is the overall defect rate when two factories contribute?_

### Step 1 — Compute each case's contribution

The picture makes the "weigh and add" step visible. We recompute the per-case contributions $P(A_i)\,P(B \mid A_i)$ and also their sum, so we have three values to plot: factory 1's share of defects, factory 2's share of defects, and the total they add up to.

```python
import numpy as np
import matplotlib.pyplot as plt

# Two factories: shares and per-factory defect rates.
pA = np.array([0.6, 0.4])
pD_given = np.array([0.02, 0.05])

# Total probability: weighted defect contributions.
contrib = pA * pD_given
vals = [contrib[0], contrib[1], contrib.sum()]
```

### Step 2 — Plot the contributions and their sum

A bar chart shows the two per-case contributions in blue and green, and the total `P(defect)` in purple. Visually the two case bars stack up to the total bar — the essence of the total probability theorem: the whole is the sum of the weighted parts.

```python
labels = ['case 1: P(A1)P(D|A1)', 'case 2: P(A2)P(D|A2)', 'total P(defect)']
colors = ['#4ea1ff', '#7ee787', '#c89bff']

plt.bar(labels, vals, color=colors)
plt.title('Total probability: weighted defect contributions sum to P(defect)')
plt.ylabel('probability')
plt.show()
```

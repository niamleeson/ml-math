# Continuous variables: PDF & CDF

_Probability & Statistics_

**When outcomes are smooth (like height), probability becomes area under a curve.**

Some quantities are continuous: height, weight, time. They can be any value in a range.

     There are infinitely many exact values, so any single one has probability 0.

## Reference implementation — NumPy + SciPy

### Step 1 — Read probabilities off the CDF

For a continuous variable, the **CDF** `F(x) = P(X <= x)` gives the probability of landing at or below `x`. For the standard normal, `F(0) = 0.5` by symmetry — half the mass sits below the mean. The probability of an *interval* is a difference of CDF values: `P(-1 < X < 1) = F(1) - F(-1)`.

```python
import numpy as np
from scipy.stats import norm

# Standard normal: PDF f(x), CDF F(x) = P(X <= x).
f_at_0 = norm.cdf(0)            # 0.5 by symmetry
p_interval = norm.cdf(1) - norm.cdf(-1)

print("F(0) = P(X<=0) =", round(f_at_0, 4))
print("P(-1 < X < 1) =", round(p_interval, 4))
```

### Step 2 — Recover the CDF by integrating the PDF

The CDF is the running **area under the PDF**. So the integral of the PDF from `-inf` up to `0` should equal `F(0)`. We approximate that integral numerically with the trapezoidal rule on a fine grid, and confirm it lands near 0.5.

```python
# Area under the PDF from -inf..0 should equal the CDF at 0.
grid = np.linspace(-8, 0, 100000)
pdf_values = norm.pdf(grid)
area = np.trapz(pdf_values, grid)

print("integral of PDF up to 0 =", round(area, 4))
```

### Step 3 — Confirm the interval probability by simulation

As a sanity check we draw a million standard-normal samples and simply count the fraction that fall in `(-1, 1)`. That empirical fraction should match the `F(1) - F(-1)` value from Step 1.

```python
# Verify the interval probability by simulation.
rng = np.random.default_rng(0)
x = rng.standard_normal(1_000_000)
in_interval = (x > -1) & (x < 1)
sim_p = np.mean(in_interval)

print("sim P(-1<X<1) =", round(sim_p, 4))
```

## Visualize the data & results

_How does the standard normal PDF relate to its CDF?_

### Step 4 — Evaluate the PDF and CDF on a grid

To plot both curves we sample them over the same range of x-values. `f(x)` is the bell-shaped density; `F(x)` is its running total, an S-shaped curve climbing from 0 to 1.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Standard normal PDF f(x) and CDF F(x) on a grid.
x = np.linspace(-4, 4, 200)
pdf = norm.pdf(x)
cdf = norm.cdf(x)
```

### Step 5 — Plot the PDF and CDF together

Drawing them on one axis makes the relationship clear: wherever the PDF is tallest (near 0) the CDF rises most steeply, and the CDF flattens out in the tails where the density is near zero.

```python
plt.plot(x, pdf, color='#4ea1ff', label='PDF f(x)')
plt.plot(x, cdf, color='#ffb454', label='CDF F(x)')
plt.title('Standard normal: PDF f(x) and CDF F(x) = P(X <= x)')
plt.xlabel('x')
plt.ylabel('density / probability')
plt.legend()
plt.show()
```

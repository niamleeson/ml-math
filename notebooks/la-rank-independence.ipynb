# Rank & linear independence

_Linear Algebra_

**How many truly independent directions a matrix's columns span.**

Vectors are linearly independent if none of them can be built from the others by scaling and adding.

     The rank of a matrix is the number of independent columns it has — the true number of directions it spans.

---

This notebook builds the idea up one step at a time: first construct a column that is secretly a combination of the others, then watch the rank reveal the redundancy, then see what independence and collapse look like geometrically.
```python
# Setup — numpy / pandas / scikit-learn / matplotlib ship with Colab.
import numpy as np
import matplotlib.pyplot as plt
```

## Reference implementation — NumPy

We build a matrix whose third column is a hidden combination of the first two, then probe it three ways: (1) measure its rank, (2) confirm the redundancy makes it singular, (3) compare against a full-rank matrix.

### Step 1 — Build a column that adds no new direction

We take two genuinely independent basis vectors `v1` and `v2`, then define `v3 = 2*v1 + 3*v2`. Because `v3` is built entirely from `v1` and `v2`, it points in a direction those two already span — it contributes nothing new. We stack all three as columns of a matrix `A`.

```python
# v3 = 2*v1 + 3*v2, so it adds no new direction.
v1 = np.array([1., 0., 0.])
v2 = np.array([0., 1., 0.])
v3 = 2 * v1 + 3 * v2

A = np.column_stack([v1, v2, v3])
print("A =\n", A)
```

### Step 2 — Measure the rank and confirm singularity

The **rank** counts independent columns. With three columns but only two independent directions, the rank is 2, not 3. A square matrix that is rank-deficient is **singular**, so its determinant is zero — a second, independent confirmation of the redundancy.

```python
# Rank counts independent columns: 2 here, even though A has 3 columns.
rank_A = np.linalg.matrix_rank(A)
print("rank:", rank_A, "of", A.shape[1], "columns")

# A dependent column makes A singular, so its determinant collapses to 0.
det_A = np.linalg.det(A)
print("det ~ 0:", np.isclose(det_A, 0.0))
```

### Step 3 — Compare against a full-rank matrix

For contrast, the 3x3 identity matrix has three columns that point along three perpendicular axes — none is a combination of the others. Its rank equals its number of columns, so it is **full rank**. This is the healthy case the dependent matrix above failed to reach.

```python
# The identity's columns are mutually independent -> full rank.
B = np.eye(3)
rank_B = np.linalg.matrix_rank(B)
print("rank(I3):", rank_B, " (full)")
```

## Visualize the data & results

_What does it look like when columns are independent versus collapsed onto a line?_

### Step 1 — Set up independent vectors and a dependent set

We pick two independent unit vectors `v1` and `v2` to show what "spanning the plane" looks like. We also build a set of vectors that all lie on a single line through the origin — a dependent set, since each is just a scaled copy of the same direction. Stacking `v1`, `v2`, and their combination confirms the rank drops to 2.

```python
# Independent basis vectors.
v1 = np.array([1., 0.])
v2 = np.array([0., 1.])

# Dependent set: every vector lies on one line through the origin.
dep = np.array([[1, 0.5], [2, 1], [3, 1.5]])

A = np.column_stack([v1, v2, 2 * v1 + 3 * v2])
print('rank', np.linalg.matrix_rank(A))    # 2 of 3
```

### Step 2 — Plot independence versus collapse

The green points are the independent vectors `v1` and `v2`, which reach in two different directions and together cover the plane. The red points are the dependent set; they all sit on the dashed red line, so they only ever explore one direction. That collapse onto a single line is exactly what a rank drop looks like geometrically.

```python
fig, ax = plt.subplots(figsize=(6, 5))

# Independent vectors reach in two distinct directions.
ax.scatter([v1[0], v2[0]], [v1[1], v2[1]], color='#7ee787',
           s=80, label='independent: v1, v2')

# Dependent vectors all sit on one line -> rank drops.
ax.scatter(dep[:, 0], dep[:, 1], color='#ff7b72',
           s=80, label='dependent (rank drops)')
ax.plot([0, 3], [0, 1.5], color='#ff7b72', linestyle='--',
        label='line every dependent vector lies on')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()

plt.tight_layout()
plt.show()
```

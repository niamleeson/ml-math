# Conditional probability

_Probability & Statistics_

**Once you learn something is true, the odds of everything else change. This measures by how much.**

New information changes the odds.

     Conditional probability answers: "given that $B$ already happened, how likely is $A$ now?"

```python
# Setup — numpy / pandas / scikit-learn / matplotlib ship with Colab.
import numpy as np
import matplotlib.pyplot as plt
```

## Reference implementation — NumPy + SciPyWe work the classic fair-die example: $A$ = "rolled a 2", $B$ = "rolled an even number". We compute $P(A \mid B)$ from the definition $P(A \mid B) = P(A \cap B)/P(B)$, then confirm it with a simulation that restricts the rolls to the even-number world. Two steps.

### Step 1 — Compute P(A | B) from the definitionOver the six equally likely faces, $B$ (even) has probability $3/6$ and $A \cap B$ (rolling a 2, which is also even) has probability $1/6$. The conditional probability is their ratio, $P(A \cap B)/P(B) = (1/6)/(3/6) = 1/3$. We build the event masks and divide.

```python
# A = "rolled a 2", B = "rolled an even number".
faces = np.arange(1, 7)
A = faces == 2
B = np.isin(faces, [2, 4, 6])

pB = np.mean(B)          # P(B)
pAB = np.mean(A & B)     # P(A and B)

print("P(A|B) = P(A and B) / P(B) =", round(pAB / pB, 4))
```

### Step 2 — Verify with a conditioned simulationConditioning on $B$ literally means: throw away every roll that isn't even, then ask how often the remaining rolls are a 2. We simulate 400,000 die rolls, keep only the even ones, and measure the fraction equal to 2. It should land near $1/3$, matching the formula.

```python
# Verify by conditioning the simulation on B.
rng = np.random.default_rng(0)
d = rng.integers(1, 7, size=400000)

# Restrict to the "even" world — these are the rolls where B happened.
given_B = d[np.isin(d, [2, 4, 6])]

print("sim P(A|B) =", round(np.mean(given_B == 2), 4))
print("expected 1/3 =", round(1 / 3, 4))
```

## Visualize the data & results_Given the die shows an even number, how likely is it a 2?_

### Step 1 — Recompute the three probabilitiesWe rebuild the event masks and compute the three quantities we want to compare: $P(B)$, $P(A \cap B)$, and the conditional $P(A \mid B) = P(A \cap B)/P(B)$. Keeping this self-contained means the plot cell runs on its own.

```python
# A = "rolled a 2", B = "rolled an even number".
faces = np.arange(1, 7)
A = faces == 2
B = np.isin(faces, [2, 4, 6])

pB = np.mean(B)
pAB = np.mean(A & B)
vals = [pB, pAB, pAB / pB]
```

### Step 2 — Plot the three probabilities side by sideThe bars show $P(B)$, $P(A \cap B)$, and $P(A \mid B)$. Notice the conditional bar is the tallest: shrinking the world to just the even rolls *raises* the chance of a 2 from $1/6$ to $1/3$. That is exactly what conditioning does — it re-normalizes within $B$.

```python
labels = ['P(B even)', 'P(A and B)', 'P(A | B)']
colors = ['#7ee787', '#c89bff', '#4ea1ff']

plt.bar(labels, vals, color=colors)
plt.title('P(A|B) = P(A and B) / P(B) for a fair die')
plt.ylabel('probability')
plt.show()
```

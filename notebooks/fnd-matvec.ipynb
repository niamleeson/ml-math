# Matrix × vector

**Apply the same dot product to every row at once.**

Multiplying a matrix by a vector does one dot product per row.

     So you can score a whole dataset of examples in a single step.

```python
# Setup — numpy / pandas / scikit-learn / matplotlib ship with Colab.
import numpy as np
import matplotlib.pyplot as plt
```

## Reference implementation — NumPy

### Step 1 — Set up the matrix and the weight vector

`A` holds two houses, one per row, each with two features `[size, bedrooms]`. The vector `x` holds one **price weight** per feature: \$200 per unit of size and \$10,000 per bedroom. Our goal is to combine them into a price for each house.

```python
import numpy as np

A = np.array([                 # two houses: [size, bedrooms]
    [1500.0, 3.0],
    [ 900.0, 2.0],
])
x = np.array([200.0, 10000.0])  # price weights
```

### Step 2 — Multiply with `A @ x`

The `@` operator takes the dot product of **each row** of `A` with `x`, producing one number per row — all houses scored in a single line. We then redo the two rows by hand to confirm `A @ x` is exactly those per-row dot products.

```python
# matrix x vector: each row of A dotted with x
Ax = A @ x
print("Ax (price per house):", Ax)        # both predictions at once

# verify it equals the per-row dot products done by hand
print("row 0 . x :", A[0] @ x)
print("row 1 . x :", A[1] @ x)
```

### Step 3 — Check the shape rule

A matrix–vector product only works when the inner dimensions match: `(m, n) @ (n,) -> (m,)`. Here `A` is `(2, 2)` and `x` is `(2,)`, so the two `n`'s agree and the result has one entry per row, shape `(2,)`.

```python
# shape rule: (m,n) @ (n,) -> (m,)  the inner n's must match
print("A.shape =", A.shape, " x.shape =", x.shape, " -> Ax.shape =", Ax.shape)
```

## Visualize the data & results

_What price does A @ x give for each house?_

### Step 1 — Recompute the per-house prices

We rebuild the same `A` and `x`, then compute `Ax = A @ x` again — one dot product per row gives the predicted price for each house, ready to plot.

```python
import numpy as np
import matplotlib.pyplot as plt

A = np.array([                 # two houses: [size, bedrooms]
    [1500.0, 3.0],
    [ 900.0, 2.0],
])
x = np.array([200.0, 10000.0])  # price weights

# matrix times vector: one dot product per row
Ax = A @ x                      # [330000, 200000]
```

### Step 2 — Plot one bar per house

A bar chart makes the two prices easy to compare. We give each bar a label describing its house, then write the exact price on top of each bar before showing the figure.

```python
fig, ax = plt.subplots()
labels = ["house 1 [1500,3]", "house 2 [900,2]"]
ax.bar(labels, Ax, color=["C0", "green"])

for i, v in enumerate(Ax):
    ax.text(i, v, str(int(v)), ha="center", va="bottom")

ax.set_ylabel("price")
ax.set_title("Ax = price per house")
plt.show()
```

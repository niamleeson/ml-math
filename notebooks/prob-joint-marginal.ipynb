# Joint & marginal distributions

_Probability & Statistics_

**Two variables at once. Sum out one to get the other back.**

Often two things vary together: height and weight, ad shown and click.

     A joint distribution gives the chance of every combination of the two.

```python
# Setup — numpy / pandas / scikit-learn / matplotlib ship with Colab.
import numpy as np
import matplotlib.pyplot as plt
```

## Reference implementation — NumPy + SciPy

### Step 1 — Build the joint distribution tableA **joint** distribution gives the probability of every (X, Y) combination at once. We store it as a 2×3 table: rows are X ∈ {0, 1} and columns are Y ∈ {0, 1, 2}. Because it is a probability distribution, all six entries must sum to 1 — we check that first.

```python
# Joint PMF table: rows = X in {0,1}, cols = Y in {0,1,2}.
joint = np.array([[0.10, 0.15, 0.05],
                  [0.20, 0.25, 0.25]])

print("joint sums to:", round(joint.sum(), 6))
```

### Step 2 — Marginalise to recover P(X) and P(Y)To get the **marginal** of one variable, you *sum out* the other. Summing each row (over Y) collapses the table to P(X); summing each column (over X) gives P(Y). In NumPy that is `axis=1` for rows and `axis=0` for columns.

```python
# Marginal of X = sum over Y (rows); marginal of Y = sum over X (cols).
pX = joint.sum(axis=1)
pY = joint.sum(axis=0)

print("P(X) =", np.round(pX, 4))
print("P(Y) =", np.round(pY, 4))
```

### Step 3 — Confirm the marginals by samplingAs a sanity check we draw many samples directly from the joint table and count how often each X and each Y value appears. We flatten the table into a 1-D list of cell probabilities, sample cell indices with those weights, then convert each flat index back to its (row, column) = (X, Y) pair with `divmod`. The empirical frequencies should match the marginals from Step 2.

```python
# Verify by sampling cells according to the joint, then counting.
rng = np.random.default_rng(0)
flat = joint.ravel()                       # 1-D list of the six cell probabilities
idx = rng.choice(flat.size, size=500000, p=flat)

# Turn each flat cell index back into its (X, Y) coordinates.
xs, ys = np.divmod(idx, joint.shape[1])

print("sim P(X) =", np.round([np.mean(xs == i) for i in range(2)], 4))
print("sim P(Y) =", np.round([np.mean(ys == j) for j in range(3)], 4))
```

## Visualize the data & results

_How do marginals fall out of a joint distribution table?_

### Step 1 — Rebuild the joint table for the plotThis visualisation cell stands on its own, so we restate the same joint table here. Each cell's value is the probability of that (X, Y) combination — the picture below will shade cells by how likely they are.

```python
# Joint PMF table: rows = X in {0,1}, cols = Y in {0,1,2}.
joint = np.array([[0.10, 0.15, 0.05],
                  [0.20, 0.25, 0.25]])
```

### Step 2 — Draw the joint as a heatmap with marginals in the labelsWe render the table as a heatmap (brighter = more probable) and write each probability into its cell. The axis tick labels carry the marginals — column sums P(Y) along the x-axis and row sums P(X) along the y-axis — so you can literally see the marginals as the row and column totals of the joint.

```python
fig, ax = plt.subplots()
im = ax.imshow(joint, cmap='viridis')

# Write each probability into its cell.
for i in range(joint.shape[0]):
    for j in range(joint.shape[1]):
        ax.text(j, i, joint[i, j], ha='center', va='center', color='w')

# Tick labels carry the marginals: column sums = P(Y), row sums = P(X).
ax.set_xticks(range(3), ['Y=0 (0.3)', 'Y=1 (0.4)', 'Y=2 (0.3)'])
ax.set_yticks(range(2), ['X=0  (sum 0.3)', 'X=1  (sum 0.7)'])
ax.set_title('Joint PMF P(X, Y); row sums = P(X), column sums = P(Y)')

fig.colorbar(im)
plt.show()
```

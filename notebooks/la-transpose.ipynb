# Transpose

_Linear Algebra_

**Flip a matrix across its diagonal: rows become columns.**

The transpose of a matrix flips it across its main diagonal.

     Row 1 becomes column 1, row 2 becomes column 2, and so on. The entry at $(i,j)$ moves to $(j,i)$.

## Reference implementation — NumPy

### Step 1 — Flip a matrix and watch its shape swap

The transpose flips a matrix across its main diagonal, so a 2×3 matrix becomes 3×2. `A.T` gives the transpose; printing both shapes shows the rows and columns have traded places.

```python
import numpy as np

A = np.array([[1, 2, 3],
              [4, 5, 6]])          # 2x3
print("A shape:", A.shape, " A.T shape:", A.T.shape)
print("A.T =\n", A.T)
```

### Step 2 — Confirm entries move from (i, j) to (j, i)

Transposing sends the entry at position `(i, j)` to position `(j, i)`. We spot-check that with one entry: `A[0, 2]` should equal `A.T[2, 0]`.

```python
# Each entry (i, j) lands at (j, i) after transposing.
print("entry swap A[0,2]==A.T[2,0]:", A[0, 2] == A.T[2, 0])
```

### Step 3 — Verify the product rule (AB)ᵀ = BᵀAᵀ

Transposing a product reverses the order of the factors: `(AB)ᵀ = BᵀAᵀ`. We build a compatible 3×2 matrix `B`, compute both sides, and confirm they are equal with `np.allclose`.

```python
# Product rule: (AB)^T == B^T A^T (note the reversed order).
B = np.array([[1, 0],
              [0, 1],
              [2, 1]])             # 3x2

lhs = (A @ B).T
rhs = B.T @ A.T
print("(AB)^T == B^T A^T:", np.allclose(lhs, rhs))
```

## Visualize the data & results

_What does flipping a matrix across its diagonal do to its shape and entries?_

### Step 1 — Build the matrix and its transpose

We take the same 2×3 matrix `A` and form its transpose `A.T`, a 3×2 matrix where every entry `(i, j)` has moved to `(j, i)`. The next step draws both so the flip is visible.

```python
import numpy as np
import matplotlib.pyplot as plt

A = np.array([[1, 2, 3],
              [4, 5, 6]])    # 2x3
At = A.T                      # 3x2: entry (i,j) -> (j,i)
```

### Step 2 — Draw both grids with their entries labelled

We render `A` and `A.T` as heatmaps and print each number in its cell. Reading across a row of `A` and down the matching column of `A.T` shows the rows-become-columns flip directly.

```python
fig, axes = plt.subplots(1, 2, figsize=(8, 4))

# Draw A on the left, A.T on the right.
for ax, M, title in zip(axes, [A, At], ['A (2x3)', 'A.T (3x2)']):
    ax.imshow(M, cmap='Blues')

    # Label every cell with its value.
    for i in range(M.shape[0]):
        for j in range(M.shape[1]):
            ax.text(j, i, str(M[i, j]), ha='center', va='center')

    ax.set_title(title)
    ax.set_xticks(range(M.shape[1]))
    ax.set_yticks(range(M.shape[0]))

plt.tight_layout()
plt.show()
```

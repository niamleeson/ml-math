# Eigenvalues & eigenvectors

_Foundations: Math you need first_

**Special directions a matrix only stretches, never rotates. The skeleton of your data.**

A matrix usually rotates and stretches a vector when it multiplies it.

     But for a few special directions, the matrix only stretches — it does not change the direction at all.

---

This notebook builds up eigenvalues and eigenvectors one step at a time. Run each cell, read the note above it, and watch how a matrix merely stretches its eigenvectors while it tilts every other direction. _Save a copy to your Drive (File → Save a copy in Drive) to edit and keep your work._

```python
# Setup — numpy / pandas / scikit-learn / matplotlib ship with Colab.
import numpy as np
import matplotlib.pyplot as plt
```

## Reference implementation — NumPy

### Step 1 — Compute the eigenvalues and eigenvectors

We start with a simple diagonal matrix `A` that scales the x-axis by 2 and the y-axis by 3. `np.linalg.eig` returns the eigenvalues (the stretch factors) and the eigenvectors as the *columns* of `vecs`. For a diagonal matrix the eigenvectors are just the coordinate axes and the eigenvalues are the diagonal entries.

```python
A = np.array([
    [2.0, 0.0],
    [0.0, 3.0],
])

# eigenvalues (lambdas) and eigenvectors (columns of vecs)
vals, vecs = np.linalg.eig(A)
print("eigenvalues :", vals)            # [2., 3.]
print("eigenvectors (columns):")
print(vecs)
```

### Step 2 — Check the defining equation A z = λ z

An eigenvector `z` is special because multiplying it by `A` only scales it: `A z` equals `λ z`, the same direction stretched by the eigenvalue. We loop over each eigenpair, compute both sides, and confirm they match with `np.allclose`.

```python
# check A z = lambda z for each eigenpair
for i in range(len(vals)):
    z = vecs[:, i]
    Az = A @ z
    lz = vals[i] * z
    match = np.allclose(Az, lz)
    print("pair %d  Az=%s  lambda*z=%s  match=%s"
          % (i, np.round(Az, 4), np.round(lz, 4), match))
```

### Step 3 — A generic direction gets tilted

To feel why eigenvectors are special, contrast them with an ordinary vector. The direction `[1, 1]` is not an eigenvector of `A`, so multiplying by `A` changes its direction (not just its length) — it tilts toward the more-stretched axis.

```python
# a generic vector gets rotated; an eigenvector does not
v = np.array([1.0, 1.0])
print("A @ [1,1] =", A @ v, "-> points a new way (not an eigenvector)")
```

## Visualize the data & results

_Does A stretch its eigenvectors without rotating them?_

### Step 1 — Send three test vectors through A

We pick two eigen-directions (`e1` along x, `e2` along y) and one generic vector `g = [1, 1]`, then apply `A` to each. The eigenvectors should come out pointing the same way, only longer; the generic vector should come out tilted.

```python
A = np.array([[2.0, 0.0], [0.0, 3.0]])
vals, vecs = np.linalg.eig(A)     # eigenvalues [2, 3]

# where A sends each test vector
e1 = np.array([1.0, 0.0])
e2 = np.array([0.0, 1.0])
g = np.array([1.0, 1.0])
Ae1 = A @ e1                       # [2, 0]
Ae2 = A @ e2                       # [0, 3]
Ag = A @ g                         # [2, 3]
```

### Step 2 — Plot the stretched vectors and stretch factors

The left panel draws where `A` sends each test vector; the dashed gray line shows `[1, 1]` *before* the matrix acts, making its tilt visible. The right panel shows the eigenvalues as bars — these are exactly the stretch factors applied along each eigen-direction.

```python
fig, (ax1, ax2) = plt.subplots(1, 2)

# left — where A sends each test vector
ax1.quiver([0, 0, 0], [0, 0, 0], [Ae1[0], Ae2[0], Ag[0]],
           [Ae1[1], Ae2[1], Ag[1]], angles="xy", scale_units="xy",
           scale=1, color=["C0", "green", "red"])
ax1.plot([0, 1], [0, 1], "--", color="gray")   # [1,1] before
ax1.set_xlim(-1, 4)
ax1.set_ylim(-1, 4)
ax1.set_xlabel("x1")
ax1.set_ylabel("x2")
ax1.set_title("Eigenvectors stretch, [1,1] tilts")

# right — eigenvalues are the stretch factors
ax2.bar(["lambda 1", "lambda 2"], np.sort(vals), color=["C0", "green"])
ax2.set_title("Eigenvalues = stretch factors")

plt.show()
```

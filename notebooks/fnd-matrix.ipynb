## Reference implementation — NumPy

### Step 1 — Build a matrix from rows of data

A matrix is just a 2-D grid: each **row** is one example (here, one house) and each **column** is one feature (size, then bedrooms). We stack three house-rows into a single NumPy array `A`. Its `.shape` tells you `(m rows, n columns)` — here `(3, 2)`.

```python
import numpy as np

# three houses (rows), two features each (size, bedrooms)
A = np.array([
    [1500.0, 3.0],
    [ 900.0, 2.0],
    [2200.0, 4.0],
])

print("matrix A:")
print(A)
print("shape (m rows, n cols):", A.shape)   # (3, 2)
```

### Step 2 — Read a single entry

To pull out one number you give **two indices**: the row first, then the column, both 0-indexed. So `A[1, 0]` is the second house's size, and `A[2, 1]` is the third house's bedroom count.

```python
# read one entry: row i, column j (0-indexed)
print("A[1, 0] house-2 size     :", A[1, 0])
print("A[2, 1] house-3 bedrooms :", A[2, 1])
```

### Step 3 — Slice a whole row or a whole column

Using `:` for one of the indices grabs everything along that axis. `A[0, :]` returns row 0 — one complete example. `A[:, 0]` returns column 0 — the same feature (size) across every house.

```python
# slice a whole row (one example) or a whole column (one feature)
print("row 0 (house 1)   :", A[0, :])
print("col 0 (all sizes) :", A[:, 0])
```

## Visualize the data & results

_How is a dataset of houses laid out as a matrix?_

### Step 1 — Draw the matrix as a colored grid

A picture makes the grid structure obvious. We re-create the same `A`, then use `imshow` to paint each cell with a color scaled to its value, and label the axes with the house and feature names.

```python
import numpy as np
import matplotlib.pyplot as plt

# three houses (rows) by two features (cols)
A = np.array([
    [1500.0, 3.0],
    [ 900.0, 2.0],
    [2200.0, 4.0],
])

fig, ax = plt.subplots()
im = ax.imshow(A, cmap="viridis", aspect="auto")
ax.set_xticks([0, 1])
ax.set_xticklabels(["size", "bedrooms"])
ax.set_yticks([0, 1, 2])
ax.set_yticklabels(["house 1", "house 2", "house 3"])
ax.set_title("Matrix A: 3 houses by 2 features")
```

### Step 2 — Write each value into its cell

The colors show the relative magnitudes, but the exact numbers are clearer as text. We loop over every row `i` and column `j` and print `A[i, j]` in the center of its cell, add a colorbar for the scale, then display the figure.

```python
# write each value into its cell
for i in range(A.shape[0]):
    for j in range(A.shape[1]):
        ax.text(j, i, str(A[i, j]), ha="center", va="center", color="white")

fig.colorbar(im)
plt.show()
```

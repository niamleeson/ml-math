# Identity & diagonal matrices

_Linear Algebra_

**The number 1 of matrices, and matrices that just scale each axis.**

The identity matrix $I$ has 1's down the diagonal and 0's everywhere else. Multiplying by it changes nothing: $AI = A$ and $Ix = x$.

     A diagonal matrix $D$ also has zeros off the diagonal, but any numbers on it. Multiplying a vector by $D$ scales each coordinate independently — stretch axis 1 by $d_1$, axis 2 by $d_2$, and so on.

## Reference implementation — NumPy

We look at the two simplest kinds of matrices: the **identity**, which leaves everything unchanged, and a **diagonal** matrix, which scales each coordinate on its own. We build it in two steps: (1) show that the identity does nothing, (2) show that a diagonal matrix is just per-axis scaling.

### Step 1 — The identity matrix does nothing

`np.eye(3)` is the 3x3 identity: 1's on the diagonal, 0's elsewhere. Multiplying any matrix `A` by it returns `A` unchanged, which is why the identity is the matrix equivalent of the number 1. We confirm `A @ I == A` with `np.allclose`.

```python
I = np.eye(3)
print("identity:\n", I)

A = np.array([[2., 5., 1.],
              [0., 3., 4.],
              [7., 1., 6.]])
print("A @ I == A:", np.allclose(A @ I, A))   # identity does nothing
```

### Step 2 — A diagonal matrix scales each axis

A diagonal matrix `D = diag(d)` has the entries of `d` on its diagonal and 0's elsewhere. Multiplying a vector `x` by `D` scales each coordinate independently: `(D x)_i = d_i · x_i`. We verify this by showing `D @ x` equals the plain element-wise product `d * x`.

```python
# diagonal matrix scales each axis on its own
d = np.array([2., 3., 5.])
D = np.diag(d)
x = np.array([4., 5., 6.])

print("D x =", D @ x)
print("d * x =", d * x)                        # same: (Dx)_i = d_i x_i
print("match:", np.allclose(D @ x, d * x))
```

## Visualize the data & results

_How does a diagonal matrix D scale a vector compared to the do-nothing identity?_

### Step 1 — Build the diagonal matrix and scale a vector

We set up `D = diag(2, 3, 5)` and the vector `x = [4, 5, 6]`, then compute `D @ x`. Each coordinate is stretched by its matching diagonal entry, giving `[8, 15, 30]`.

```python
d = np.array([2., 3., 5.])
D = np.diag(d)                  # diag(2, 3, 5)
x = np.array([4., 5., 6.])
scaled = D @ x                  # 8, 15, 30
```

### Step 2 — Plot the matrix and the per-axis scaling

The left panel shows `D` as a labelled heatmap — bright on the diagonal, dark off it. The right panel is a bar chart of the scaled coordinates, making it clear that each axis was stretched by its own factor (×2, ×3, ×5) and nothing mixes between axes.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(9, 4))

# Left — D as a labelled heatmap
ax1.imshow(D, cmap='Blues')
for i in range(3):
    for j in range(3):
        ax1.text(j, i, str(int(D[i, j])), ha='center', va='center')
ax1.set_title('D = diag(2, 3, 5)')

# Right — each coordinate after its own scaling
labels = ['x1 times 2', 'x2 times 3', 'x3 times 5']
ax2.bar(labels, scaled, color=['#4ea1ff', '#7ee787', '#ffb454'])
for i, v in enumerate(scaled):
    ax2.text(i, v, str(int(v)), ha='center', va='bottom')
ax2.set_title('D scales each coordinate')

plt.tight_layout()
plt.show()
```

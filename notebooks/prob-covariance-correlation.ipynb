# Covariance & correlation

_Probability & Statistics_

**Do two variables move together? These numbers tell you, and by how much.**

When one variable goes up, does the other tend to go up too?

     Covariance answers that: positive means they rise together, negative means one rises as the other falls.

```python
# Setup — numpy / pandas / scikit-learn / matplotlib ship with Colab.
import numpy as np
import matplotlib.pyplot as plt
```

## Reference implementation — NumPy + SciPy

### Step 1 — Build two related variables

We draw `X` from a standard normal, then build `Y` to *depend* on `X`: `Y = 2X + noise`. Because `Y` moves with `X`, we expect a positive covariance. The added noise keeps the relationship from being perfect, which is what we'll see reflected in the correlation.

```python
from scipy.stats import pearsonr

rng = np.random.default_rng(0)
n = 200000

X = rng.normal(0, 1, size=n)
Y = 2 * X + rng.normal(0, 1, size=n)   # Y depends on X, plus independent noise
```

### Step 2 — Measure covariance and correlation

`np.cov` returns a 2×2 matrix; the off-diagonal entry `[0, 1]` is `Cov(X, Y)`. Covariance has units (and scale) of `X·Y`, so it's hard to read on its own. Correlation rescales it to the range `[-1, 1]`, and `np.corrcoef` and `scipy.stats.pearsonr` should agree.

```python
# Covariance matrix: the off-diagonal entry is Cov(X, Y).
cov = np.cov(X, Y)
print("Cov(X,Y) =", round(cov[0, 1], 4))

# Correlation rescales covariance to [-1, 1].
corr = np.corrcoef(X, Y)[0, 1]
print("corrcoef =", round(corr, 4))

# scipy.stats.pearsonr gives the same correlation (plus a p-value we ignore).
r, _ = pearsonr(X, Y)
print("scipy pearson r =", round(r, 4))
```

### Step 3 — Check against the theory

Since `Y = 2X + noise` with `Var(X) = 1`, the algebra gives `Cov(X, Y) = 2·Var(X) = 2`. The correlation works out to `2 / sqrt(2²·Var(X) + Var(noise)) = 2 / sqrt(5) ≈ 0.894`. Compare this expected value to the empirical numbers above — they should match closely.

```python
# Theory: Cov = 2*Var(X) = 2; corr = 2 / sqrt(2^2*Var(X) + Var(noise)).
expected_corr = 2 / np.sqrt(5)
print("expected corr ~", round(expected_corr, 4))
```

## Visualize the data & results

_What do positive, negative, and zero correlation look like?_

### Step 1 — Build three relationships from one X

We use a single `X` and pair it with three different `Y`s: one that rises with `X` (positive), one that falls as `X` rises (negative), and one drawn independently (near zero). Sharing the same `X` makes the contrast between the three patterns easy to see.

```python
# One shared X; three Y's: positive, negative, and zero correlation.
rng = np.random.default_rng(0)
X = rng.normal(0, 1, size=40)

Y_pos = 2 * X + rng.normal(0, 1, size=40)    # rises with X
Y_neg = -2 * X + rng.normal(0, 1, size=40)   # falls as X rises
Y_zero = rng.normal(0, 1, size=40)           # independent of X
```

### Step 2 — Plot all three on one scatter

Overlaying the three clouds shows the shape behind each correlation value: an upward band (positive), a downward band (negative), and a shapeless blob (zero). Correlation is, visually, how tightly the points hug a line.

```python
plt.scatter(X, Y_pos, color='#7ee787', label='Y = 2X + noise  (r=0.91)')
plt.scatter(X, Y_neg, color='#ff7b72', label='Y = -2X + noise  (r<0)')
plt.scatter(X, Y_zero, color='#9aa7b4', label='Y independent of X  (r~0)')
plt.title('Three samples: positive (r=0.91), negative, and zero correlation')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

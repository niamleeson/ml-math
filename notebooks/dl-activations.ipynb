# Activation functions

_Deep Learning_

**The squish that lets a network learn curvy, non-straight patterns.**

An activation function takes the neuron's number $z$ and reshapes it.

     Why bother? Without it, stacking layers just makes one big straight line. Straight lines cannot learn curvy patterns.

```python
# Setup — numpy / pandas / scikit-learn / matplotlib ship with Colab.
import numpy as np
import matplotlib.pyplot as plt
```

## Reference implementation — NumPy

We'll feed a spread of input values through the four most common activation functions and compare how each one reshapes them. We build it in two steps: (1) pick the inputs, (2) apply each activation and read off the results.

### Step 1 — Choose a spread of pre-activation values

`z` holds the raw numbers a neuron might output *before* an activation is applied — a mix of negatives, zero, and positives. Using a spread like this lets us see exactly how each function treats small, zero, and large inputs.

```python
z = np.array([-3.0, -1.0, 0.0, 2.0, 5.0])   # try a spread of values
print("z", z)
```

### Step 2 — Apply each activation and compare

Each function reshapes `z` differently: **sigmoid** squashes everything into (0, 1), **tanh** into (-1, 1), **ReLU** zeros out negatives and keeps positives, and **leaky ReLU** keeps a tiny slope for negatives so they don't fully vanish. Compare the rows below to feel how each curve bends the same inputs.

```python
sigmoid = 1.0 / (1.0 + np.exp(-z))          # squish into (0, 1)
tanh = np.tanh(z)                           # squish into (-1, 1)
relu = np.maximum(0.0, z)                   # keep positives, zero negatives
leaky = np.where(z > 0, z, 0.01 * z)        # tiny slope for negatives

print("z      ", z)
print("sigmoid", np.round(sigmoid, 3))
print("tanh   ", np.round(tanh, 3))
print("relu   ", relu)
print("leaky  ", leaky)
```

## Visualize the data & results

_For a real digit-classifier neuron's pre-activation z, how does each activation reshape it?_

### Step 1 — Sample the activation curves over a range of z

To draw smooth curves we evaluate each function at many evenly spaced `z` values from -5 to 5. A real digit-classifier neuron in this example produced a pre-activation around 1.63, which we'll mark on the plot for context.

```python
# the real digit-classifier neuron produced z around 1.6; show the full curve
z = np.linspace(-5, 5, 31)

sigmoid = 1.0 / (1.0 + np.exp(-z))      # squish into (0, 1)
tanh = np.tanh(z)                       # squish into (-1, 1)
relu = np.maximum(0.0, z)               # keep positives, zero negatives
```

### Step 2 — Plot the curves

Plotting the three curves together makes their shapes obvious: sigmoid and tanh flatten (saturate) at the extremes, while ReLU grows without bound on the positive side. The dashed vertical line marks the real neuron's `z` so you can read off what each activation would output there.

```python
plt.plot(z, sigmoid, color="#4ea1ff", label="sigmoid (0,1)")
plt.plot(z, tanh, color="#7ee787", label="tanh (-1,1)")
plt.plot(z, relu, color="#c89bff", label="ReLU max(0,z)")
plt.axvline(1.63, color="gray", linestyle="--")   # this neuron's real z
plt.title("Activation functions over a neuron's z (digit classifier)")
plt.xlabel("z (pre-activation)")
plt.ylabel("activation(z)")
plt.legend()
plt.show()
```

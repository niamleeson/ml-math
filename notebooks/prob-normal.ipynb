## Reference implementation — NumPy + SciPy

### Step 1 — Define a Normal distribution and check the 68-95-99.7 rule

We model IQ-like scores as `Normal(mu=100, sigma=15)`. SciPy's `norm` object lets us ask for the **CDF** — the probability of landing at or below a value. The probability of falling *within* `k` standard deviations of the mean is `CDF(mu + k·sigma) - CDF(mu - k·sigma)`. For a Normal those come out to roughly 68%, 95%, and 99.7% — the famous empirical rule.

```python
import numpy as np
from scipy.stats import norm

mu = 100
sigma = 15
d = norm(loc=mu, scale=sigma)

# 68-95-99.7 rule: probability within 1, 2, 3 standard deviations.
for k in (1, 2, 3):
    upper = d.cdf(mu + k * sigma)
    lower = d.cdf(mu - k * sigma)
    p = upper - lower
    print("within %d sigma: %.4f" % (k, p))
```

### Step 2 — Standardize to z-scores and confirm by simulation

**Standardizing** rescales any Normal into the *standard* Normal: `z = (x - mu) / sigma`. After this transform the mean is 0 and the standard deviation is 1. We draw a million samples, standardize them, and check that the empirical z-mean is near 0 and z-std near 1 — and that about 68% of the z-scores fall within one standard deviation, matching Step 1.

```python
# Standardizing: z = (x - mu) / sigma is standard normal.
rng = np.random.default_rng(0)
x = rng.normal(mu, sigma, size=1_000_000)
z = (x - mu) / sigma

z_mean = z.mean()
z_std = z.std()
frac_within_1 = np.mean(np.abs(z) < 1)

print("z mean ~0:", round(z_mean, 4), " z std ~1:", round(z_std, 4))
print("sim within 1 sigma:", round(frac_within_1, 4))
```

## Visualize the data & results

_Does the 68-95-99.7 rule hold for Normal(100, 15)?_

### Step 3 — Compute the bell curve and the within-k-sigma bars

To draw the picture we need two things: the PDF (height of the bell) across a grid of x-values, and the three within-`k`-sigma probabilities we computed earlier. We gather both here before plotting.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Normal(100, 15) PDF, and the within-k-sigma probabilities.
mu = 100
sigma = 15
d = norm(loc=mu, scale=sigma)

x = np.linspace(40, 160, 200)
pdf = d.pdf(x)

within = []
for k in (1, 2, 3):
    p = d.cdf(mu + k * sigma) - d.cdf(mu - k * sigma)
    within.append(p)
```

### Step 4 — Plot the PDF beside the empirical-rule bars

Left: the smooth bell curve of `Normal(100, 15)`. Right: the probability mass captured within 1, 2, and 3 standard deviations — three bars that should sit at roughly 0.68, 0.95, and 0.997, confirming the rule visually.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(11, 4))

# Left — the bell curve.
ax1.plot(x, pdf, color='#4ea1ff')
ax1.set_title('Normal(mu=100, sigma=15) PDF')

# Right — probability within k standard deviations.
labels = ['within 1 sigma', 'within 2 sigma', 'within 3 sigma']
colors = ['#7ee787', '#ffb454', '#c89bff']
ax2.bar(labels, within, color=colors)
ax2.set_title('68-95-99.7 rule: probability within k standard deviations')

plt.show()
```

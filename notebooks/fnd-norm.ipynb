# Norms (the length of a vector)

**How big is a vector? Norms measure that. They power 'distance' and regularization.**

A norm measures the size (length) of a vector — a single non-negative number.

     The most common one is ordinary straight-line length, from the Pythagorean theorem.

## Reference implementation — NumPy

### Step 1 — Measure length two ways

The **L2 norm** is ordinary straight-line length: the square root of the sum of squares (Pythagoras). The **L1 norm** instead adds up absolute values. `np.linalg.norm` computes L2 by default, and L1 when you pass `ord=1`. For `x = [3, -4]` these come out to 5 and 7.

```python
import numpy as np

x = np.array([3.0, -4.0])

# L2 = sqrt(sum of squares) = Pythagoras
l2 = np.linalg.norm(x)            # default ord=2
# L1 = sum of absolute values
l1 = np.linalg.norm(x, ord=1)
print("x   :", x)
print("L2  :", l2, " (= sqrt(9+16) = 5)")
print("L1  :", l1, " (= |3|+|-4| = 7)")
```

### Step 2 — Reproduce the norms by hand

To see exactly what `norm` does, we rebuild each one from primitives. L2 is `sqrt(sum(x**2))`, and L1 is `sum(abs(x))`. These match the values from Step 1.

```python
# do it by hand to see what norm() computes
print("L2 by hand:", np.sqrt(np.sum(x ** 2)))
print("L1 by hand:", np.sum(np.abs(x)))
```

### Step 3 — Turn a norm into a distance

The **distance** between two vectors is just the norm of their difference, `||a - b||`. For `a = [0, 6, 8]` and the origin `b`, the difference is `a` itself, so the L2 distance is `sqrt(36 + 64) = 10`.

```python
# distance between two vectors = norm of their difference
a = np.array([0.0, 6.0, 8.0])
b = np.array([0.0, 0.0, 0.0])
print("||a - b|| :", np.linalg.norm(a - b))   # = 10
```

## Visualize the data & results

_How big is x = [3, -4], and how far apart are two vectors?_

### Step 1 — Compare the two norms as bars

We recompute the L2 and L1 norms of `x = [3, -4]`, then draw them side by side as bars. This shows at a glance that the L1 sum (7) is larger than the L2 straight-line length (5) for this vector.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([3.0, -4.0])
l2 = np.linalg.norm(x)          # 5.0, straight-line length
l1 = np.linalg.norm(x, ord=1)   # 7.0, sum of absolute values

fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.bar(["L2 norm", "L1 norm"], [l2, l1], color=["C0", "orange"])
ax1.set_title("Two sizes of x = [3, -4]")
```

### Step 2 — Draw x as a right triangle

The L2 norm is the hypotenuse of a right triangle whose legs are the components of `x`. On the right axes we draw `x` as an arrow, then dash in the horizontal leg (3) and the vertical leg (-4); the arrow's length is the L2 norm, 5.

```python
# x as a right triangle: hypotenuse is the L2 length
ax2.quiver(0, 0, x[0], x[1], angles="xy", scale_units="xy", scale=1, color="C0")
ax2.plot([0, x[0]], [0, 0], "--", color="orange")        # leg a = 3
ax2.plot([x[0], x[0]], [0, x[1]], "--", color="purple")  # leg b = -4
ax2.set_xlim(-1, 4)
ax2.set_ylim(-5, 1)
ax2.set_xlabel("x1")
ax2.set_ylabel("x2")
ax2.set_title("x as a right triangle, L2 = " + str(l2))
plt.show()
```

## Reference implementation — NumPy + SciPyWe count the ways to pick $r=3$ items from $n=5$: **permutations** (order matters) and **combinations** (order doesn't). We get the closed-form counts from formulas, confirm them by brute-force enumeration, and finally check the relationship between the two. Three steps.

### Step 1 — Counts from the closed-form formulasThe permutation count is $P(n,r) = n!/(n-r)!$ and the combination count is $C(n,r) = n!/(r!\,(n-r)!)$. SciPy's `perm` and `comb` evaluate these directly. Permutations exceed combinations because each unordered pick can be arranged in several orders.

```python
from scipy.special import perm, comb

n, r = 5, 3
print("P(n,r) = n!/(n-r)! =", int(perm(n, r)))
print("C(n,r) = n!/(r!(n-r)!) =", int(comb(n, r)))
```

### Step 2 — Confirm by brute-force enumerationRather than trust the formulas, we list every arrangement. `itertools.permutations` yields every ordered selection and `itertools.combinations` every unordered one; counting them should reproduce the numbers from Step 1 exactly.

```python
from itertools import permutations, combinations

items = list(range(n))

# Enumerate every ordered and every unordered selection of r items.
n_perm = len(list(permutations(items, r)))
n_comb = len(list(combinations(items, r)))

print("enumerated permutations:", n_perm)
print("enumerated combinations:", n_comb)
```

### Step 3 — Relate permutations to combinationsEvery combination of $r$ items can be ordered in $r!$ different ways, so $P(n,r) = C(n,r) \times r!$. Dividing the permutation count by the combination count should therefore give exactly $r!$. We check both sides match.

```python
from math import factorial

# Each combination orders into r! permutations.
ratio = n_perm // n_comb
print("perm / comb =", ratio, " r! =", factorial(r))
```

## Visualize the data & results_Picking 3 from 5: how many ordered arrangements vs unordered picks?_

### Step 1 — Gather the three countsWe collect the three numbers we want to compare for $n=5$, $r=3$: the ordered count $P(5,3)$, the unordered count $C(5,3)$, and the factor $r! = 3!$ that connects them. Keeping this self-contained lets the plot cell run on its own.

```python
from scipy.special import perm, comb
from math import factorial

# Pick r from n: ordered (permutations) vs unordered (combinations).
n, r = 5, 3
vals = [int(perm(n, r)), int(comb(n, r)), factorial(r)]
```

### Step 2 — Plot ordered vs unordered countsThe bars show $P(5,3)$, $C(5,3)$, and $r!$. The ordered bar is exactly $r!$ times taller than the unordered bar — a visual statement of $P(n,r) = C(n,r)\times r!$, the cost of caring about order.

```python
labels = ['P(5,3) ordered', 'C(5,3) unordered', 'r! = 3!']
colors = ['#4ea1ff', '#7ee787', '#ffb454']

plt.bar(labels, vals, color=colors)
plt.title('Permutations vs combinations (n=5, r=3)')
plt.ylabel('count')
plt.show()
```

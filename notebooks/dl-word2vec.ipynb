**Learn word vectors by predicting nearby words. Then king − man + woman ≈ queen.**

How do we get good embeddings? word2vec learns them by a guessing game.

     In the skip-gram version, given a center word, the model tries to predict the words around it. To do that well, it must put related words near each other.

```python
# Setup — numpy / pandas / scikit-learn / matplotlib ship with Colab.
import numpy as np
import matplotlib.pyplot as plt
```

## Reference implementation — NumPy

### Step 1 — Lay out vectors on two interpretable axes

To see the analogy clearly we hand-build 2-D vectors on a *royalty* axis and a *gender* axis. "king" is high-royalty / high-gender(male); "woman" is low on both. Because the axes carry meaning, vector arithmetic between words will carry meaning too.

```python
import numpy as np

# tiny hand-built embeddings (royalty axis, gender axis)
vec = {
    "king":  np.array([0.9, 0.9]),
    "man":   np.array([0.1, 0.9]),
    "woman": np.array([0.1, 0.1]),
    "queen": np.array([0.9, 0.1]),
}
```

### Step 2 — Compute the analogy vector

The classic analogy is king − man + woman. Subtracting "man" strips out the male component and adding "woman" swaps in the female one, while the royalty component is left untouched. The result should land wherever "queen" lives.

```python
result = vec["king"] - vec["man"] + vec["woman"]   # the analogy
print("king - man + woman =", result)
```

### Step 3 — Find the nearest word by cosine similarity

Cosine similarity measures the angle between two vectors, ignoring their length, so it captures "point in the same direction" = "similar meaning". We compare the analogy result to every candidate (excluding the input words) and pick the closest — which should be "queen".

```python
# find the closest word by cosine similarity
def cos(a, b):
    return a @ b / (np.linalg.norm(a) * np.linalg.norm(b))

candidates = [w for w in vec if w not in ("king", "man", "woman")]
best = max(candidates, key=lambda w: cos(result, vec[w]))
print("nearest word:", best)
```

## Visualize the data & results

_Is the nearest word to (king - man + woman) actually 'queen', using real word vectors?_

### Step 1 — Set up richer, higher-dimensional vectors

Now we use 4-D vectors for a wider word set, including distractors like "princess" and "cat". We store them as NumPy arrays so we can do the same king − man + woman arithmetic, but this time against more candidates to make sure "queen" really wins.

```python
import numpy as np
import matplotlib.pyplot as plt

vec = {"king": [0.9, 0.95, 0.1, 0.8], "queen": [0.88, 0.1, 0.12, 0.82],
       "princess": [0.83, 0.12, 0.18, 0.72], "man": [0.2, 0.92, 0.3, 0.1],
       "woman": [0.18, 0.08, 0.32, 0.12], "cat": [0.1, -0.85, -0.7, 0.3]}
vec = {k: np.array(v) for k, v in vec.items()}
```

### Step 2 — Score every candidate against the analogy

We compute the analogy vector again, then measure its cosine similarity to each candidate word. A higher score means the candidate points more in the same direction as king − man + woman.

```python
result = vec["king"] - vec["man"] + vec["woman"]   # the analogy vector

def cos(a, b):
    return a @ b / (np.linalg.norm(a) * np.linalg.norm(b))

cands = ["queen", "princess", "king", "woman", "cat"]
sims = [cos(result, vec[w]) for w in cands]
```

### Step 3 — Plot the similarities

A bar chart of the scores makes the winner obvious: "queen" should be the tallest bar, beating near-misses like "princess" and unrelated words like "cat". That is the analogy king − man + woman ≈ queen, confirmed by the numbers.

```python
colors = ["#7ee787", "#4ea1ff", "#ffb454", "#4ea1ff", "#ff7b72"]

plt.bar(cands, sims, color=colors)
plt.title("Cosine similarity of (king - man + woman) to candidate words")
plt.show()
```

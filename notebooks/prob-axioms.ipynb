# Probability axioms

_Probability & Statistics_

**Three simple rules that every probability must obey. No more, no less.**

A probability is a number between 0 and 1 attached to an event.

     0 means "never happens". 1 means "always happens".

This notebook checks the three axioms one at a time on a fair die — non-negativity, total mass one, and additivity for disjoint events — then confirms additivity against a simulation. ## Reference implementation — NumPy + SciPy

### Step 1 — Axioms 1 and 2: non-negativity and total mass one

Every face of a fair die gets probability `1/6`. The first axiom says no probability is negative; the second says the probabilities over the whole sample space `Omega` sum to exactly `1`.

We build the probability vector and check both at once.

```python
import numpy as np

# Fair die: probability of each face.
faces = np.arange(1, 7)
p = np.full(6, 1 / 6)

all_nonneg = bool(np.all(p >= 0))      # Axiom 1: every probability >= 0
total_mass = p.sum()                   # Axiom 2: P(Omega) = 1
print("all >= 0:", all_nonneg, " sum P(Omega) =", round(total_mass, 6))
```

### Step 2 — Axiom 3: additivity, and the complement rule

For **disjoint** events (no shared outcomes), the third axiom says `P(A or B) = P(A) + P(B)`. Take `A = {1, 2}` and `B = {5, 6}` — they share no faces, so adding their probabilities should equal the probability of their union.

The complement rule `P(not A) = 1 - P(A)` follows directly from Axioms 2 and 3, and we verify it too.

```python
# Disjoint events A={1,2}, B={5,6}: additivity.
A = np.isin(faces, [1, 2])
B = np.isin(faces, [5, 6])
pA = p[A].sum()
pB = p[B].sum()
pAorB = p[A | B].sum()
print("P(A)+P(B) =", round(pA + pB, 4), " P(A or B) =", round(pAorB, 4))

# Complement rule: P(not A) = 1 - P(A).
print("1 - P(A) =", round(1 - pA, 4), " P(A^c) =", round(p[~A].sum(), 4))
```

### Step 3 — Confirm additivity by simulation

Theory says `P(A or B) = 1/3 + 1/3 = 2/3 ≈ 0.667`. We roll the die 300,000 times and measure the fraction of rolls landing in `A or B = {1, 2, 5, 6}`. The simulated frequency should sit very close to the predicted value, showing the axiom holds empirically.

```python
# Simulate additivity.
rng = np.random.default_rng(0)
d = rng.integers(1, 7, size=300000)
sim_p = np.mean(np.isin(d, [1, 2, 5, 6]))
print("sim P(A or B) =", round(sim_p, 4))
```

## Visualize the data & results

_Do the three axioms hold for disjoint die events A={1,2} and B={5,6}?_

### Step 1 — Recompute the five quantities to plot

To picture additivity, we recompute the same probabilities as a list: `P(A)`, `P(B)`, their sum, the union `P(A or B)`, and the complement `P(not A)`. Bars for `P(A)+P(B)` and `P(A or B)` should match exactly — that visual equality *is* the additivity axiom.

```python
import numpy as np
import matplotlib.pyplot as plt

# Fair die; disjoint events A={1,2}, B={5,6}.
faces = np.arange(1, 7)
p = np.full(6, 1 / 6)
A = np.isin(faces, [1, 2])
B = np.isin(faces, [5, 6])
pA = p[A].sum()
pB = p[B].sum()
vals = [pA, pB, pA + pB, p[A | B].sum(), p[~A].sum()]
```

### Step 2 — Draw the bar chart

Plot the five quantities. The third bar (`P(A)+P(B)`) and the fourth (`P(A or B)`) reach the same height, confirming additivity visually.

```python
labels = ['P(A)', 'P(B)', 'P(A)+P(B)', 'P(A or B)', 'P(not A)']
colors = ['#4ea1ff', '#7ee787', '#c89bff', '#c89bff', '#ffb454']
plt.bar(labels, vals, color=colors)
plt.title('Additivity: P(A)+P(B) equals P(A or B)')
plt.ylabel('probability')
plt.show()
```

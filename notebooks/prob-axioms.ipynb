_Probability & Statistics_

**Three simple rules that every probability must obey. No more, no less.**

A probability is a number between 0 and 1 attached to an event.

     0 means "never happens". 1 means "always happens".

## Reference implementation — NumPy + SciPy

```python
import numpy as np

# Fair die: probability of each face.
faces = np.arange(1, 7)
p = np.full(6, 1 / 6)
print("all >= 0:", bool(np.all(p >= 0)), " sum P(Omega) =", round(p.sum(), 6))

# Disjoint events A={1,2}, B={5,6}: additivity.
A = np.isin(faces, [1, 2]); B = np.isin(faces, [5, 6])
pA, pB = p[A].sum(), p[B].sum()
pAorB = p[A | B].sum()
print("P(A)+P(B) =", round(pA + pB, 4), " P(A or B) =", round(pAorB, 4))

# Complement rule: P(not A) = 1 - P(A).
print("1 - P(A) =", round(1 - pA, 4), " P(A^c) =", round(p[~A].sum(), 4))

# Simulate additivity.
rng = np.random.default_rng(0)
d = rng.integers(1, 7, size=300000)
print("sim P(A or B) =", round(np.mean(np.isin(d, [1, 2, 5, 6])), 4))
```

## Visualize the data & results

_Do the three axioms hold for disjoint die events A={1,2} and B={5,6}?_

```python
import numpy as np
import matplotlib.pyplot as plt

# Fair die; disjoint events A={1,2}, B={5,6}.
faces = np.arange(1, 7)
p = np.full(6, 1 / 6)
A = np.isin(faces, [1, 2])
B = np.isin(faces, [5, 6])
pA, pB = p[A].sum(), p[B].sum()
vals = [pA, pB, pA + pB, p[A | B].sum(), p[~A].sum()]

labels = ['P(A)', 'P(B)', 'P(A)+P(B)', 'P(A or B)', 'P(not A)']
colors = ['#4ea1ff', '#7ee787', '#c89bff', '#c89bff', '#ffb454']
plt.bar(labels, vals, color=colors)
plt.title('Additivity: P(A)+P(B) equals P(A or B)')
plt.ylabel('probability')
plt.show()
```
